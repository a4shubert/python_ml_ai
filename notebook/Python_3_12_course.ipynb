# 1. Basic Data Types and Core Values


This chapter refreshes the Python values that show up everywhere. For an educated adult returning to Python, the important questions are not only what each type stores, but whether it is mutable, ordered, hashable, exact, approximate, sequence-like, or structurally meaningful.

Much of Python fluency comes from choosing the right type for the job and understanding what that choice communicates to readers.


| Type | Ordered | Mutable | Hashable | Typical use |
|---|---|---|---|---|
| `int` / `float` / `complex` | n/a | No | Yes | Numeric computation |
| `bool` | n/a | No | Yes | Conditions and logic |
| `str` | Yes | No | Yes | Text |
| `bytes` | Yes | No | Yes | Binary payloads |
| `bytearray` | Yes | Yes | No | Mutable binary data |
| `list` | Yes | Yes | No | Evolving sequences |
| `tuple` | Yes | No | Usually | Fixed records |
| `set` / `frozenset` | No | `set`: Yes / `frozenset`: No | `frozenset`: Yes | Uniqueness and membership |
| `dict` | In insertion order | Yes | No | Named structure and lookup |
| `None` | n/a | No | Yes | Missing / absent value |


## 1.1 `int`, `float`, `complex`


`int` is exact for whole numbers, `float` is approximate for real-valued computation, and `complex` handles numbers with real and imaginary parts. The key refresher point is that the type is part of the meaning, not just storage.


```python
whole_number = 42
decimal_number = 3.75
complex_number = 2 + 3j

print(type(whole_number), whole_number)
print(type(decimal_number), decimal_number)
print(type(complex_number), complex_number)
```

```python
x = 0.1 + 0.2
print("0.1 + 0.2 =", x)
print("Equals 0.3 exactly?", x == 0.3)
print("Magnitude of 2 + 3j:", abs(2 + 3j))
```

## 1.2 `bool` and truthiness


`bool` has only two values, but truthiness extends the idea to empty containers, zero values, and objects that define their own truth semantics. Real Python code relies on that constantly.


```python
print(True, False)
print(type(True))
print("True + True + False =", True + True + False)
```

```python
samples = [0, 1, "", "python", [], [1], None]
print([(value, bool(value)) for value in samples])
```

## 1.3 `str` and Unicode text


`str` stores Unicode text and behaves like an immutable sequence. The main habits to refresh are slicing, formatting, normalization by simple methods, and awareness that Python text is Unicode by default.


```python
course_name = "Python 3.12 Course"
print(course_name.lower())
print(course_name[:6])
print("Python" in course_name)
```

```python
raw_line = "  alice,ML engineer,uk  "
name, role, country = [part.strip() for part in raw_line.split(",")]
print(f"{name.title()} works as a {role} in {country.upper()}.")
```

## 1.4 `bytes` and `bytearray`


`bytes` is an immutable sequence of integers from 0 to 255, while `bytearray` is mutable. They matter whenever code crosses into files, networks, encodings, or binary protocols.


```python
payload = b"ABC"
print(payload)
print(list(payload))
print(payload.decode("ascii"))
```

```python
mutable_payload = bytearray(b"ABC")
mutable_payload[0] = ord("Z")
print(mutable_payload)
print(bytes(mutable_payload))
```

## 1.5 `list`


`list` is the default mutable sequence. It is flexible and convenient, but its very flexibility means it should be chosen intentionally rather than automatically.


```python
topics = ["syntax", "functions", "numpy"]
topics.append("pandas")
topics[0] = "basic syntax"
print(topics)
```

```python
scores = [88, 92, 79, 95]
curved_scores = [min(score + 3, 100) for score in scores if score >= 80]
print(scores)
print(curved_scores)
```

## 1.6 `tuple`


`tuple` is an immutable ordered collection. It is most useful when values belong together as a fixed record or when immutability itself is part of the meaning.


```python
rgb = (34, 139, 34)
print(rgb)
print(rgb[1])
```

```python
student_record = ("Alice", 91, "passed")
name, score, status = student_record
lookup = {(1, 2): "point A"}
print(name, score, status)
print(lookup[(1, 2)])
```

## 1.7 `set` and `frozenset`


`set` emphasizes uniqueness and fast membership checks. `frozenset` is the immutable counterpart and can itself be used as a dictionary key or set element.


```python
tags = {"python", "ml", "python", "ai"}
print(tags)
print("ml" in tags)
```

```python
completed = frozenset({"variables", "loops", "functions"})
planned = frozenset({"functions", "numpy", "pandas"})
print(completed & planned)
print(completed | planned)
```

## 1.8 `dict`


`dict` is one of Python’s most expressive and important types. It models named structure, preserves insertion order, and underlies much of Python’s configuration and JSON-shaped data.


```python
student = {"name": "Alex", "score": 93, "track": "Python 3.12"}
print(student)
print(student["name"])
```

```python
scores = {"alice": 91, "bob": 84, "cara": 95}
passed = {name: score for name, score in scores.items() if score >= 90}
print(scores)
print(passed)
```

## 1.9 `None`, identity, and sentinels


`None` means “no meaningful value here”. The key semantic point is that `None` is distinct from `0`, `False`, or an empty collection, and should be compared with `is`, not `==`.


```python
result = None
print(result is None)
print(type(result))
```

```python
def find_even(numbers):
    for number in numbers:
        if number % 2 == 0:
            return number
    return None

print(find_even([1, 3, 5]) is None)
print(find_even([1, 4, 5]))
```

## 1.10 Mutability, hashability, and value vs reference behavior


Many Python bugs are really about object identity and mutation. This section should make the learner alert to aliasing, in-place mutation, and the fact that only hashable objects can be dictionary keys or set elements.


```python
numbers = [1, 2, 3]
alias = numbers
alias.append(4)
print(numbers)
print(numbers is alias)
```

```python
immutable_key = ("course", 1)
lookup = {immutable_key: "intro"}
print(lookup[immutable_key])
try:
    { [1, 2]: "bad key" }
except TypeError as exc:
    print(type(exc).__name__, exc)
```

# 2. Expressions, Operators, and Control Flow


This chapter shifts from values to execution. The goal is to refresh how Python binds names, evaluates expressions, chooses branches, iterates, and encodes common control-flow patterns in a way that is explicit and readable.

A large amount of Python fluency comes from seeing that control flow is not just about syntax. It is about evaluation order, truthiness, short-circuiting, iterator consumption, and choosing a style that communicates intent clearly.


| Topic | Key question | Why it matters |
|---|---|---|
| Assignment | What does a name refer to? | Python binds names, not variables in the C-like sense |
| Operators | How is an expression evaluated? | Correctness depends on precedence, identity, and membership semantics |
| Branching | Which path runs and why? | Truthiness and matching shape readable decisions |
| Iteration | What is being consumed? | Loops often hide iterator semantics |
| Comprehensions | What are we transforming? | Expressive, but easy to over-compress |


## 2.1 Names, binding, and assignment


`=` in Python binds a name to an object; it does not copy values by default. This is a subtle but central idea, especially for mutable objects and function arguments.


```python
value = 10
alias = value
print(value, alias)

items = [1, 2]
other_name = items
other_name.append(3)
print(items, other_name)
```

```python
x, y = 1, 2
x, y = y, x
print(x, y)

config = {"mode": "draft"}
backup = config
backup["mode"] = "final"
print(config)
```

## 2.2 Arithmetic, comparison, logical, membership, and identity operators


Operators in Python are expressive but semantically distinct. A refresher should emphasize the difference between value comparison, object identity, membership, and logical composition.


```python
a = 7
b = 3
print(a + b, a - b, a * b, a / b, a // b, a % b, a ** b)
```

```python
name = "python"
other_name = "".join(["py", "thon"])
print(name == other_name)
print(name is other_name)
print("py" in name)
print((3 < 5) and (5 < 8))
```

## 2.3 Conditional expressions and short-circuiting


Conditional expressions let you express small decisions inline. Short-circuiting means logical operators stop once the result is already determined, which affects both efficiency and side effects.


```python
score = 72
status = "pass" if score >= 50 else "fail"
print(status)
```

```python
def expensive_check():
    print("expensive check ran")
    return True

print(False and expensive_check())
print(True or expensive_check())
```

## 2.4 `if` / `elif` / `else`


Branching is where truthiness meets explicit decision-making. The main design question is not how to branch, but how to make branch logic readable and unsurprising.


```python
temperature = 18
if temperature < 10:
    label = "cold"
elif temperature < 20:
    label = "mild"
else:
    label = "warm"
print(label)
```

```python
user_input = ""
if user_input:
    print("received:", user_input)
else:
    print("input was empty")
```

## 2.5 `for` and `while`


`for` expresses iteration over an iterable, while `while` expresses repetition under a condition. In Python, `for` is often preferred because it is safer and more declarative.


```python
for index, letter in enumerate("abc", start=1):
    print(index, letter)
```

```python
countdown = 3
while countdown > 0:
    print(countdown)
    countdown -= 1
print("done")
```

## 2.6 `break`, `continue`, `pass`


These statements alter loop behavior directly. The main nuance is to use them where they simplify control flow, not where they make the path through a loop harder to reconstruct mentally.


```python
for number in range(1, 6):
    if number == 4:
        break
    print(number)
```

```python
for number in range(1, 6):
    if number % 2 == 0:
        continue
    print("odd:", number)

if True:
    pass
```

## 2.7 `match` / `case`


`match` adds structural pattern matching to Python. It is most useful when code is branching on shape or category rather than just boolean conditions.


```python
command = ("move", 10, 20)

match command:
    case ("move", x, y):
        print(f"moving to {x}, {y}")
    case ("quit",):
        print("quitting")
    case _:
        print("unknown command")
```

```python
status_code = 404
match status_code:
    case 200:
        print("ok")
    case 400 | 404:
        print("client-side issue")
    case _:
        print("other")
```

## 2.8 Comprehensions and generator expressions


Comprehensions are concise transformation tools; generator expressions provide lazy evaluation. The refresh point is that these are not just shorter loops, they encode a particular style of thinking.


```python
numbers = range(8)
squares_of_even = [n * n for n in numbers if n % 2 == 0]
print(squares_of_even)
```

```python
numbers = range(8)
lazy_squares = (n * n for n in numbers if n % 2 == 0)
print(lazy_squares)
print(list(lazy_squares))
```

## 2.9 Common control-flow idioms


Experienced Python code often uses compact patterns such as `any`, `all`, `enumerate`, sentinel checks, and guard clauses. These idioms compress intent without obscuring the logic.


```python
values = [2, 4, 6, 8]
print(any(v > 5 for v in values))
print(all(v % 2 == 0 for v in values))
```

```python
records = ["alpha", "beta", "gamma"]
for index, record in enumerate(records, start=1):
    print(f"{index}: {record}")
```

# 3. Functions and Call Semantics


Functions are where Python starts to scale intellectually. This chapter is about abstraction, call semantics, scope, and designing interfaces that are precise but still Pythonic.

A strong refresher should make function definitions feel like interface design rather than just code packaging. Names, defaults, argument kinds, return shapes, and docstrings all shape how other people can use the code.


| Topic | Key question | Why it matters |
|---|---|---|
| Parameters | What may callers supply, and how? | Function signatures are the public contract |
| Scope | Where does a name resolve? | Prevents subtle bugs and improves predictability |
| Closures | What state is retained? | Useful for specialization and lightweight abstraction |
| First-class functions | Can behavior be passed around cleanly? | Enables functional and callback styles |
| API design | Is the function easy to understand and hard to misuse? | Readability and correctness both depend on this |


## 3.1 Defining functions


A function definition creates a reusable unit of behavior and a public calling interface. Good function definitions usually do one thing clearly and expose only the parameters the caller should care about.


```python
def greet(name):
    return f"Hello, {name}"

print(greet("Alice"))
```

```python
def area(width, height):
    return width * height

print(area(5, 3))
```

## 3.2 Parameters, arguments, defaults


Parameters live in the function definition; arguments are the values provided by the caller. Default values are convenient, but they encode policy and should be chosen deliberately.


```python
def power(base, exponent=2):
    return base ** exponent

print(power(5))
print(power(5, 3))
```

```python
def connect(host, port=5432, secure=True):
    return {"host": host, "port": port, "secure": secure}

print(connect("db.internal"))
```

## 3.3 Positional-only and keyword-only arguments


Modern Python lets you control how arguments must be passed. This is useful when you want to stabilize APIs, avoid ambiguous calls, or make intent explicit at the call site.


```python
def ratio(a, b, /):
    return a / b

print(ratio(10, 2))
```

## 3.7 Closures and `nonlocal`


A closure lets an inner function retain access to values from its enclosing scope. `nonlocal` allows rebinding in that enclosing scope, which is useful for stateful lightweight abstractions.


```python
def make_multiplier(factor):
    def multiply(value):
        return value * factor
    return multiply

double = make_multiplier(2)
print(double(8))
```

```python
def make_counter():
    count = 0
    def increment():
        nonlocal count
        count += 1
        return count
    return increment

counter = make_counter()
print(counter(), counter(), counter())
```

## 3.8 First-class functions

## 3.9 Docstrings, signatures, and API design


A function signature is part of the API, and a docstring is part of the contract with the reader. Good Python APIs are explicit about what they require, what they return, and what behavior is guaranteed.


```python
import inspect

def normalize_name(name: str, /, *, upper: bool = False) -> str:
    """Normalize a personal name for display."""
    cleaned = name.strip()
    return cleaned.upper() if upper else cleaned.title()

print(inspect.signature(normalize_name))
print(normalize_name.__doc__)
```

# 4. Python’s Data Model


Python’s data model is the language’s hidden spine. It explains why objects can be iterable, callable, comparable, printable, or usable in a `with` block simply by satisfying the relevant protocol.

A graduate-level refresher should leave you seeing Python syntax less as a bag of features and more as a coherent object protocol system. Many advanced-looking behaviors are just normal objects implementing a few special methods.


| Topic | Core idea | Why it matters |
|---|---|---|
| Objects and types | Everything is an object with a type | Explains uniform behavior across the language |
| Dunder methods | Special methods define protocols | Custom classes integrate with built-ins naturally |
| Truth, length, containment | Objects can define how they behave in core operations | Makes APIs feel native |
| Iteration and representation | Objects can participate in loops and display clearly | Enables idiomatic user-defined abstractions |
| Hashing and equality | Identity and value semantics are separate design choices | Critical for dictionaries, sets, and correctness |


## 4.1 Objects, types, and classes


In Python, values, functions, classes, modules, and even exceptions are all objects. Every object has a type, and user-defined classes are themselves objects too.


```python
value = 3.14
print(type(value))
print(type(len))
print(type(str))
```

```python
class Student:
    pass

student = Student()
print(type(student))
print(type(Student))
```

## 4.2 Dunder methods and protocol-based design


Special methods such as `__len__`, `__iter__`, and `__repr__` let objects participate in Python protocols. This is the mechanism behind much of Python’s apparent flexibility.


```python
class Playlist:
    def __init__(self, songs):
        self.songs = songs

    def __len__(self):
        return len(self.songs)

playlist = Playlist(["Intro", "Theme", "Credits"])
print(len(playlist))
```

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    def __repr__(self):
        return f"Temperature({self.celsius}C)"

print(Temperature(21))
```

## 4.3 Truth-value, length, and containment protocols


Python asks objects how they behave in boolean contexts, length checks, and membership tests. Built-ins use these protocols automatically, which is why custom objects can feel native.


```python
class Basket:
    def __init__(self, items):
        self.items = items

    def __len__(self):
        return len(self.items)

basket = Basket(["apple", "banana"])
print(bool(basket))
print(len(basket))
```

```python
class Vocabulary:
    def __init__(self, words):
        self.words = set(words)

    def __contains__(self, item):
        return item.lower() in self.words

vocab = Vocabulary({"python", "typing", "async"})
print("Python" in vocab)
print("rust" in vocab)
```

## 4.4 Iteration protocol


Iteration in Python is driven by the iterator protocol: `iter(obj)` should produce an iterator, and that iterator should yield items via `next()` until it raises `StopIteration`.


```python
class Countdown:
    def __init__(self, start):
        self.current = start

    def __iter__(self):
        return self

    def __next__(self):
        if self.current == 0:
            raise StopIteration
        self.current -= 1
        return self.current + 1

print(list(Countdown(3)))
```

```python
items = iter(["a", "b", "c"])
print(next(items))
print(next(items))
print(list(items))
```

## 4.5 Representation and string conversion


Python distinguishes between developer-facing representation (`repr`) and user-facing display (`str`). Good custom objects usually implement at least `__repr__` clearly.


```python
class Report:
    def __init__(self, title):
        self.title = title

    def __repr__(self):
        return f"Report(title={self.title!r})"

    def __str__(self):
        return f"Report: {self.title}"

report = Report("Quarterly Review")
print(repr(report))
print(str(report))
```

```python
data = {"name": "Alice", "score": 95}
print(repr(data))
print(str(data))
```

## 4.6 Equality, ordering, and hashing


Equality, ordering, and hashing are related but distinct choices. If two objects compare equal and are immutable, they should normally produce the same hash to behave correctly in sets and dictionaries.


```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __eq__(self, other):
        return isinstance(other, Point) and (self.x, self.y) == (other.x, other.y)

p1 = Point(1, 2)
p2 = Point(1, 2)
print(p1 == p2)
print(p1 is p2)
```

```python
pairs = [(2, "b"), (1, "c"), (1, "a")]
print(sorted(pairs))
print(hash((1, 2, 3)))
```

## 4.7 Context manager protocol


Context managers let objects control setup and cleanup around a block of code. This protocol is what makes `with` safe and expressive for resources.


```python
class DemoContext:
    def __enter__(self):
        print("entering")
        return "resource"

    def __exit__(self, exc_type, exc, tb):
        print("leaving")

with DemoContext() as resource:
    print("using", resource)
```

```python
from pathlib import Path

path = Path("context_demo.txt")
with path.open("w") as f:
    _ = f.write("hello")

with path.open() as f:
    print(f.read())
```

## 4.8 Callables and descriptor preview


Anything implementing `__call__` can behave like a function. Descriptors are a deeper mechanism behind methods, properties, and attribute access control, and are worth previewing even before a full deep dive.


```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, value):
        return value * self.factor

double = Multiplier(2)
print(double(10))
```

```python
class Example:
    @property
    def label(self):
        return "descriptor-backed property"

print(Example().label)
```

## 4.9 Why Python feels “duck typed”


Python often cares more about supported behavior than declared ancestry. If an object behaves like the thing needed, many APIs are satisfied without requiring a specific concrete type.


```python
def total_lengths(items):
    return sum(len(item) for item in items)

print(total_lengths([[1, 2], "abc", (10, 20, 30)]))
```

```python
class DataChunk:
    def __len__(self):
        return 5

chunk = DataChunk()
print(len(chunk))
print(bool(chunk))
```

### Common Mistake / Nuance

Python is not “magic” here: most flexible behavior comes from protocols, not from special treatment of custom classes. If an object behaves unexpectedly in a loop, comparison, or boolean context, inspect which protocol it is actually implementing.

# 6. Error Handling and Defensive Programming


Reliable Python code uses exceptions, assertions, and cleanup mechanisms deliberately. This chapter should make defensive programming feel normal rather than ceremonial.

A useful refresher perspective is that error handling is not an afterthought layered on top of working code. It is part of the interface design: what can fail, how failure is communicated, and how invariants are protected.


| Topic | Core question | Why it matters |
|---|---|---|
| Exceptions | What counts as failure? | Python communicates many failures through exceptions |
| Structured handling | Where should errors be caught? | Correct boundaries improve both safety and readability |

| Assertions and invariants | What must always be true? | Prevents silent corruption of program state |

## 6.1 Exceptions as control and error signals


Exceptions signal that normal execution cannot continue in the current path. In Python they are used not only for catastrophic problems, but also for ordinary failures such as missing keys, invalid input, or exhausted iterators.


```python
try:
    int("not-a-number")
except ValueError as exc:
    print(type(exc).__name__, exc)
```

```python
items = iter([1])
print(next(items))
try:
    print(next(items))
except StopIteration:
    print("iterator exhausted")
```

## 6.2 `try` / `except` / `else` / `finally`


Python’s exception-handling block gives four distinct control points. `except` handles failure, `else` runs only on success, and `finally` runs regardless, which makes it useful for cleanup.


```python
try:
    result = 10 / 2
except ZeroDivisionError:
    print("division failed")
else:
    print("success:", result)
finally:
    print("finished attempt")
```

```python
for denominator in [2, 0]:
    try:
        print(10 / denominator)
    except ZeroDivisionError:
        print("cannot divide by zero")
```

## 6.3 Raising exceptions well


Raising an exception is part of API design. The exception type and message should explain both what failed and why that failure matters to the caller.


```python
def parse_positive_int(text: str) -> int:
    value = int(text)
    if value <= 0:
        raise ValueError("value must be positive")
    return value

print(parse_positive_int("12"))
```

```python
def require_non_empty(text: str) -> str:
    cleaned = text.strip()
    if not cleaned:
        raise ValueError("text must not be empty after stripping")
    return cleaned

try:
    require_non_empty("   ")
except ValueError as exc:
    print(exc)
```

## 6.4 Exception hierarchy and custom exceptions


Python exceptions form a hierarchy, which means callers can choose narrow or broad handling. Custom exception classes are useful when domain failures deserve names of their own.


```python
class ConfigurationError(Exception):
    pass

def load_config(config):
    if "host" not in config:
        raise ConfigurationError("missing host")
    return config

try:
    load_config({})
except ConfigurationError as exc:
    print(type(exc).__name__, exc)
```

```python
try:
    raise FileNotFoundError("demo")
except OSError as exc:
    print("caught through superclass:", type(exc).__name__)
```

## 6.5 Assertions and invariants


Assertions are for conditions that should never be false if the program is correct. They are not a substitute for validating user input, but they are useful for defending internal assumptions.


```python
def mean(values):
    assert values, "values must not be empty"
    return sum(values) / len(values)

print(mean([10, 20, 30]))
```

```python
items = [1, 2, 3]
assert all(isinstance(item, int) for item in items)
print("all items are integers")
```

## 6.6 Resource safety


Defensive programming includes ensuring that resources such as files, sockets, or locks are released even when operations fail. Python’s `with` statement is the standard way to express that intent.


```python
from pathlib import Path

path = Path("resource_demo.txt")
with path.open("w") as f:
    _ = f.write("safe write")

with path.open() as f:
    print(f.read())
```

```python
class Resource:
    def __enter__(self):
        print("acquire")
        return self

    def __exit__(self, exc_type, exc, tb):
        print("release")

with Resource():
    print("using resource")
```

## 6.7 EAFP vs LBYL


Python often prefers EAFP: “easier to ask forgiveness than permission.” That means trying the operation and handling failure, rather than checking every precondition up front. But LBYL can still be clearer in some contexts.


```python
data = {"name": "Alice"}
try:
    print(data["score"])
except KeyError:
    print("score missing")
```

```python
data = {"name": "Alice"}
if "score" in data:
    print(data["score"])
else:
    print("score missing")
```

## 6.8 Writing diagnosable code


Diagnosable code fails clearly, names its assumptions, and preserves useful context. Good error-handling style is not only about preventing crashes; it is about making failures understandable.


```python
def load_user(record):
    try:
        return record["name"].strip().title()
    except KeyError as exc:
        raise KeyError("record is missing required field 'name'") from exc

try:
    load_user({})
except KeyError as exc:
    print(exc)
```

```python
def divide(a, b):
    if b == 0:
        raise ZeroDivisionError(f"cannot divide {a} by zero")
    return a / b

try:
    divide(10, 0)
except ZeroDivisionError as exc:
    print(exc)
```

### Common Mistake / Nuance

Catching exceptions too broadly often hides programming mistakes and makes debugging harder. Catch the narrowest exception that expresses the failure you expect, and raise domain-specific errors when that improves clarity for callers.

# 9. Object-Oriented Python


Object-oriented Python is useful when state and behavior genuinely belong together. This chapter should emphasize pragmatic OO design rather than class-heavy ritual.

A strong refresher should leave the learner able to tell when a class clarifies a design, when a dataclass is enough, and when functions plus plain data would be simpler.


| Topic | Core question | Why it matters |
|---|---|---|
| Class design | What state and behavior belong together? | Prevents accidental complexity |
| Variables and methods | What lives on the class versus the instance? | Clarifies shared vs local behavior |
| Inheritance | Are we specializing behavior or just sharing code? | Avoids fragile hierarchies |
| Encapsulation | What should callers rely on? | Helps APIs stay stable |
| Dataclasses and protocols | What is the lightest abstraction that works? | Python rewards restraint |


## 9.1 Class definition and instance state


A class defines the behavior shared by instances; each instance holds its own state. The key design question is whether the state and operations naturally belong together.


```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

alice = Student("Alice", 95)
bob = Student("Bob", 84)
print(alice.name, alice.score)
print(bob.name, bob.score)
```

```python
class Task:
    def __init__(self, title, done=False):
        self.title = title
        self.done = done

    def complete(self):
        self.done = True

task = Task("Write notebook")
task.complete()
print(task.title, task.done)
```

## 9.2 Class variables vs instance variables


Class variables are shared across instances; instance variables belong to each object separately. Confusing the two leads to subtle state bugs and misleading APIs.


```python
class BankAccount:
    interest_rate = 0.02

    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

a = BankAccount("Alice", 100)
b = BankAccount("Bob", 200)
print(a.interest_rate, b.interest_rate)
```

```python
a.balance += 50
BankAccount.interest_rate = 0.03
print(a.balance, b.balance)
print(a.interest_rate, b.interest_rate)
```

## 9.3 Methods: instance, class, static


Python supports instance methods, class methods, and static methods, each for a different purpose. Use them intentionally rather than decoratively.


```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    def to_fahrenheit(self):
        return self.celsius * 9 / 5 + 32

print(Temperature(20).to_fahrenheit())
```

```python
class User:
    def __init__(self, name):
        self.name = name

    @classmethod
    def guest(cls):
        return cls("guest")

    @staticmethod
    def is_valid_name(name):
        return bool(name.strip())

print(User.guest().name)
print(User.is_valid_name(" Alice "))
```

## 9.4 Inheritance and composition


Inheritance models “is-a” specialization, while composition builds larger behavior from smaller parts. In Python, composition is often the more robust default.


```python
class Animal:
    def speak(self):
        return "..."

class Dog(Animal):
    def speak(self):
        return "woof"

print(Dog().speak())
```

```python
class Engine:
    def start(self):
        return "engine started"

class Car:
    def __init__(self):
        self.engine = Engine()

    def start(self):
        return self.engine.start()

print(Car().start())
```

## 9.5 Method resolution order


When inheritance becomes layered, Python resolves attributes through the method resolution order (MRO). Understanding this is important even if you rarely build deep hierarchies yourself.


```python
class A:
    def who(self):
        return "A"

class B(A):
    pass

print(B().who())
print(B.__mro__)
```

```python
class Base:
    def label(self):
        return "base"

class Derived(Base):
    def label(self):
        return "derived -> " + super().label()

print(Derived().label())
```

## 9.6 Encapsulation by convention


Python relies more on naming conventions than on strict access restrictions. Single underscores signal non-public intent; double underscores trigger name mangling but are not true privacy.


```python
class Account:
    def __init__(self, owner, balance):
        self.owner = owner
        self._balance = balance

account = Account("Alice", 100)
print(account.owner)
print(account._balance)
```

```python
class SecretBox:
    def __init__(self):
        self.__token = "hidden"

box = SecretBox()
print(hasattr(box, "__token"))
print(box._SecretBox__token)
```

## 9.7 `dataclasses`


`dataclasses` reduce boilerplate for classes whose primary role is to store structured data. They are often the right starting point when a dict feels too loose but a custom class feels repetitive.


```python
from dataclasses import dataclass

@dataclass
class StudentRecord:
    name: str
    score: int

print(StudentRecord("Alice", 95))
```

```python
from dataclasses import dataclass

@dataclass(order=True)
class LeaderboardEntry:
    score: int
    name: str

entries = [LeaderboardEntry(88, "Bob"), LeaderboardEntry(95, "Alice")]
print(sorted(entries, reverse=True))
```

## 9.8 Protocol-oriented vs inheritance-heavy design


In Python, objects often work because they support the right operations, not because they inherit from a particular base class. This usually leads to simpler and more flexible designs.


```python
def render_all(items):
    for item in items:
        print(item.render())

class HtmlReport:
    def render(self):
        return "<h1>report</h1>"

class TextReport:
    def render(self):
        return "report"

render_all([HtmlReport(), TextReport()])
```

```python
class JsonReport:
    def render(self):
        return '{"report": true}'

print(JsonReport().render())
```

## 9.9 When OOP is a good fit in Python


OOP is a good fit when you have evolving state plus behavior that naturally belongs to that state. It is a poor fit when a few small functions over plain data would already express the design more clearly.


```python
class Timer:
    def __init__(self):
        self.ticks = 0

    def tick(self):
        self.ticks += 1
        return self.ticks

timer = Timer()
print(timer.tick(), timer.tick())
```

```python
scores = [91, 84, 95]
print(sum(scores) / len(scores))
print("This calculation does not need a class.")
```

### Common Mistake / Nuance

Python does not reward class design for its own sake. Many weak Python designs come from building inheritance hierarchies where a dataclass, a protocol-like interface, or plain functions over simple data would be clearer.

# 10. Functional and Declarative Styles


Python supports functional and declarative styles, but usually rewards moderation. The key theme is expressing transformations cleanly without sacrificing readability.

A strong refresher should make it clear that Python is not a purely functional language, but it gives you powerful functional tools: first-class functions, closures, iterator pipelines, higher-order utilities, and a bias toward expressing what should happen rather than narrating every step.


| Topic | Core question | Why it matters |
|---|---|---|
| Functions as values | Can behavior be passed and stored cleanly? | Enables flexible composition |
| Higher-order tools | When should `map`, `filter`, or `reduce` replace loops? | Helps choose the right abstraction |
| Closures and partials | How can behavior be specialized without new classes? | Supports lightweight reusable design |
| Immutability | When is it useful to avoid mutation? | Reduces incidental state coupling |
| Readability | Is the abstraction making the code clearer? | Functional style fails when it becomes opaque |


## 10.1 Functions as values


In Python, functions are ordinary objects: they can be assigned, stored, passed, and returned. This matters because it makes behavior composable in the same way that data is composable.


```python
def shout(text):
    return text.upper()

transform = shout
print(transform("python"))
```

```python
def apply(function, value):
    return function(value)

print(apply(str.title, "machine learning"))
```

## 10.2 `map`, `filter`, `reduce`


These tools express common transformation patterns, but they are not always the clearest option. In Python, list comprehensions often compete directly with `map` and `filter` on readability.


```python
numbers = [1, 2, 3, 4]
print(list(map(lambda x: x * 10, numbers)))
print(list(filter(lambda x: x % 2 == 0, numbers)))
```

```python
from functools import reduce

values = [1, 2, 3, 4]
product = reduce(lambda a, b: a * b, values)
print(product)
```

## 10.3 Lambdas: when useful, when not


Lambdas are small anonymous functions. They are useful when behavior is truly local and simple, but they become hard to read when they hold substantial logic.


```python
pairs = [("alice", 91), ("bob", 84), ("cara", 95)]
print(sorted(pairs, key=lambda item: item[1], reverse=True))
```

## 10.4 Closures and partial application


Closures retain state from their enclosing scope, and partial application pre-fills some arguments of a function. Both are useful for producing specialized behavior without introducing new classes.


```python
def make_power(exponent):
    def power(base):
        return base ** exponent
    return power

square = make_power(2)
print(square(5))
```

```python
from functools import partial

def format_name(name, *, uppercase=False):
    return name.upper() if uppercase else name.title()

format_loud = partial(format_name, uppercase=True)
print(format_loud("alice"))
```

## 10.5 Immutability as a design tool


Python does not force immutability, but immutable design often reduces accidental coupling. The main value is not ideology; it is making state changes explicit and easier to reason about.


```python
numbers = (1, 2, 3)
updated_numbers = numbers + (4,)
print(numbers)
print(updated_numbers)
```

```python
settings = {"theme": "classic", "autosave": True}
updated = {**settings, "autosave": False}
print(settings)
print(updated)
```

## 10.6 Iterator algebra with `itertools`


`itertools` is one of Python’s clearest bridges into functional style. It gives you composable iterator-building tools that support lazy, pipeline-oriented programming.

## 10.7 Functional style in idiomatic Python


Idiomatic Python uses functional ideas selectively. The goal is usually not maximal abstraction, but a clean and direct transformation pipeline that stays easy to read.


```python
records = [
    {"name": "Alice", "score": 91},
    {"name": "Bob", "score": 84},
    {"name": "Cara", "score": 95},
]
high_scores = [record["name"] for record in records if record["score"] >= 90]
print(high_scores)
```

```python
text = ["  python", "async  ", "  typing  "]
normalized = [item.strip().upper() for item in text]
print(normalized)
```

## 10.8 Readability tradeoffs


Functional style is helpful only while it improves clarity. A compact pipeline can be elegant, but once the reader has to mentally simulate too much structure, the abstraction has probably become counterproductive.


```python
ugly = list(filter(lambda x: x % 2 == 0, map(lambda x: x * 2, range(6))))
clear = [x * 2 for x in range(6) if x % 2 == 0]

print("compressed style:", ugly)
print("clearer style:", clear)
```

```python
def transformed(values):
    result = []
    for value in values:
        if value % 2 == 0:
            result.append(value * 2)
    return result

print(transformed(range(6)))
```

### Common Mistake / Nuance

Functional style is not automatically better style. In Python, the best version is usually the one that makes transformation logic obvious without forcing the reader to decode a dense chain of abstractions.

Some interfaces accept more than one type, some values may be absent, and some parameters should be restricted to a fixed set of valid options. These typing tools express those cases directly.

## 12.4 `Protocol` and structural typing


Protocols support structural typing: an object is compatible if it provides the required behavior, regardless of inheritance. This aligns naturally with Python’s duck-typed design.


```python
from typing import Protocol

class Renderer(Protocol):
    def render(self) -> str: ...

class HtmlReport:
    def render(self) -> str:
        return "<h1>report</h1>"

def show(item: Renderer) -> None:
    print(item.render())

show(HtmlReport())
```

```python
from typing import Protocol

class SupportsLen(Protocol):
    def __len__(self) -> int: ...

def size_of(item: SupportsLen) -> int:
    return len(item)

print(size_of([1, 2, 3]))
print(size_of("python"))
```

## 12.7 Static typing as documentation and design aid


Typing is not only for tooling. It often improves the design itself by forcing the author to decide what a function consumes, what it returns, and which cases are intentionally allowed.

Type hints do not automatically validate runtime data, prevent misuse from untyped callers, or fully model every dynamic pattern Python allows. They are helpful, but not omnipotent.

## 13.3 `threading`

## 13.5 `async` / `await`


Async code expresses cooperative concurrency. Coroutines suspend explicitly at `await` points, which makes the control flow structured but different from threaded code.


```python
import asyncio

async def fetch(name: str, delay: float) -> str:
    await asyncio.sleep(delay)
    return f"finished {name}"

async def main():
    print(await fetch("a", 0.01))

asyncio.run(main())
```

```python
import asyncio

async def pair():
    results = await asyncio.gather(fetch("x", 0.01), fetch("y", 0.01))
    print(results)

asyncio.run(pair())
```

## 13.6 Event loops and tasks


The event loop is the scheduler behind async execution. Tasks wrap coroutines so that multiple operations can make progress together under loop control.


```python
import asyncio

async def worker(name):
    await asyncio.sleep(0.01)
    return name.upper()

async def orchestrate():
    task1 = asyncio.create_task(worker("alpha"))
    task2 = asyncio.create_task(worker("beta"))
    print(await task1, await task2)

asyncio.run(orchestrate())
```

## 13.7 Async iteration and async context managers


Async code extends the usual iteration and context-manager ideas into awaitable forms. These patterns matter in real async libraries even if the syntax appears less often in small scripts.


```python
import asyncio

class AsyncCounter:
    def __init__(self, limit):
        self.current = 0
        self.limit = limit

    def __aiter__(self):
        return self

    async def __anext__(self):
        if self.current >= self.limit:
            raise StopAsyncIteration
        self.current += 1
        await asyncio.sleep(0.001)
        return self.current

async def consume():
    values = []
    async for item in AsyncCounter(3):
        values.append(item)
    print(values)

asyncio.run(consume())
```

```python
import asyncio

class AsyncResource:
    async def __aenter__(self):
        print("async open")
        return self

    async def __aexit__(self, exc_type, exc, tb):
        print("async close")

async def use_resource():
    async with AsyncResource():
        print("inside async context")

asyncio.run(use_resource())
```

Choosing the right model is a design decision, not a syntax preference. Threads, processes, and coroutines each fit different latency, throughput, and complexity profiles.

```python
def choose_model(workload):
    if workload == "cpu":
        return "processes"
    if workload == "network":
        return "async or threads"
    return "it depends on coordination and simplicity"

print(choose_model("cpu"))
print(choose_model("network"))
```

### Common Mistake / Nuance

Concurrency models are not interchangeable. Async is not a drop-in replacement for threads, and threads are not a substitute for process-based parallelism. The right choice depends on whether the work is CPU-bound, I/O-bound, coordination-heavy, or latency-sensitive.

## 14.1 `getattr`, `setattr`, `hasattr`

## 14.3 Decorators


Decorators wrap functions or classes with reusable behavior. They are powerful because they let cross-cutting concerns such as logging, timing, caching, or validation be attached declaratively.


```python
def emphasize(func):
    def wrapper(*args, **kwargs):
        return f"**{func(*args, **kwargs)}**"
    return wrapper

@emphasize
def greet(name):
    return f"hello {name}"

print(greet("alice"))
```

```python
def log_call(func):
    def wrapper(*args, **kwargs):
        print(f"calling {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

@log_call
def square(x):
    return x * x

print(square(5))
```

## 14.4 Descriptors


Descriptors are objects that customize attribute access through `__get__`, `__set__`, and `__delete__`. They are one of the mechanisms behind methods, properties, and some framework APIs.


```python
class UpperCase:
    def __get__(self, instance, owner):
        return instance._name.upper()

    def __set__(self, instance, value):
        instance._name = value

class Person:
    name = UpperCase()

    def __init__(self, name):
        self.name = name

p = Person("alice")
print(p.name)
```

```python
class Number:
    def __set_name__(self, owner, name):
        self.private_name = "_" + name

    def __get__(self, instance, owner):
        return getattr(instance, self.private_name)

    def __set__(self, instance, value):
        setattr(instance, self.private_name, int(value))

class Counter:
    count = Number()

    def __init__(self, count):
        self.count = count

print(Counter(3.8).count)
```

## 14.5 Properties


Properties expose method logic through attribute syntax. They are especially useful when you want validation or derived behavior while keeping the public API simple.


```python
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius

    @property
    def fahrenheit(self):
        return self._celsius * 9 / 5 + 32

print(Temperature(20).fahrenheit)
```

```python
class Score:
    def __init__(self, value):
        self.value = value

    @property
    def value(self):
        return self._value

    @value.setter
    def value(self, value):
        if value < 0:
            raise ValueError("score cannot be negative")
        self._value = value

print(Score(95).value)
```

## 14.6 Metaclasses overview


Metaclasses control class creation itself. Most Python programmers rarely need them directly, but understanding that classes are objects created by a metaclass helps demystify some advanced frameworks.


```python
class MyClass:
    pass

print(type(MyClass))
print(type(int))
```

```python
class RegistryMeta(type):
    registry = []

    def __new__(mcls, name, bases, namespace):
        cls = super().__new__(mcls, name, bases, namespace)
        mcls.registry.append(name)
        return cls

class Demo(metaclass=RegistryMeta):
    pass

print(RegistryMeta.registry)
```

## 14.8 Writing powerful but readable Python


Advanced Python should still feel readable. The highest-level skill is restraint: using introspection and metaprogramming only when they reduce duplication or clarify interfaces more than they increase mental load.


```python
def call_if_present(obj, method_name):
    method = getattr(obj, method_name, None)
    if callable(method):
        return method()
    return None

class Greeter:
    def greet(self):
        return "hello"

print(call_if_present(Greeter(), "greet"))
```

The most common mistake with advanced Python features is using them because they are possible rather than because they are necessary. If a dynamic trick makes the code harder to inspect than the explicit version, it is usually the wrong tradeoff.
