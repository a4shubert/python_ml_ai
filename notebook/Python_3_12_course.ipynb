## 1.1 `pathlib`


`pathlib` gives you an object-oriented way to work with files and folders. It is easier to read than string-based path code and works well across operating systems.


```python
from pathlib import Path

current_dir = Path.cwd()

print("Current working directory:", current_dir)
print("Notebook files:", [p.name for p in current_dir.glob("*.ipynb")])
```

## 1.2 `datetime`


`datetime` is the standard tool for representing dates, times, and time differences. It is essential for timestamps, scheduling, and simple date arithmetic.


```python
from datetime import datetime

now = datetime.now()
print("Current timestamp:", now)
print("Formatted:", now.strftime("%Y-%m-%d %H:%M"))
```

```python
from datetime import date, timedelta

course_start = date(2026, 5, 1)
week_later = course_start + timedelta(days=7)

print("Course start:", course_start)
print("One week later:", week_later)
print("Days between:", (week_later - course_start).days)
```

## 1.4 `csv`


`csv` is useful when working with simple table data stored as comma-separated text. It helps you read and write rows without manual string splitting.


```python
import csv
from pathlib import Path

csv_path = Path("stdlib_examples") / "scores.csv"

with csv_path.open("w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["name", "score"])
    writer.writerow(["Alice", 92])
    writer.writerow(["Bob", 88])

print(csv_path.read_text())
```

```python
import csv
from pathlib import Path

csv_path = Path("stdlib_examples") / "scores.csv"

with csv_path.open() as f:
    reader = csv.DictReader(f)
    rows = list(reader)

print(rows)
print("Average score:", sum(int(row["score"]) for row in rows) / len(rows))
```

## 1.5 `collections.Counter`


`Counter` counts hashable items for you. It is a fast way to build frequency tables, rank common values, and summarize repeated events.


```python
from collections import Counter

words = "python standard library makes everyday tasks easier".split()
word_counts = Counter(words)
print(word_counts)
```

```python
from collections import Counter

grades = ["A", "B", "A", "C", "B", "A", "B"]
grade_counts = Counter(grades)

print("Most common:", grade_counts.most_common(2))
print("Number of A grades:", grade_counts["A"])
```

## 1.6 `collections.defaultdict`


`defaultdict` creates missing values automatically. It is especially useful for grouping, counting, and building nested structures.


```python
from collections import defaultdict

scores_by_student = defaultdict(list)
for name, score in [("Alice", 90), ("Bob", 84), ("Alice", 95), ("Bob", 88)]:
    scores_by_student[name].append(score)

print(dict(scores_by_student))
```

```python
from collections import defaultdict

grouped_words = defaultdict(set)
for word in ["apple", "apricot", "banana", "blueberry"]:
    grouped_words[word[0]].add(word)

print({letter: sorted(words) for letter, words in grouped_words.items()})
```

## 1.7 `collections.deque`


`deque` is a double-ended queue optimized for fast appends and pops from both sides. It is a good fit for queues, sliding windows, and history buffers.


```python
from collections import deque

tasks = deque(["load data", "clean data"])
tasks.append("train model")
tasks.appendleft("open notebook")

print(tasks)
print("Next task:", tasks.popleft())
print("Remaining:", tasks)
```

```python
from collections import deque

recent_scores = deque(maxlen=3)
for score in [70, 80, 90, 95, 100]:
    recent_scores.append(score)
    print(list(recent_scores))
```

## 1.8 `itertools`


`itertools` provides efficient iterator building blocks. These tools help you generate combinations, repeated patterns, and streamed transformations without creating unnecessary lists.


```python
from itertools import combinations

features = ["age", "income", "city"]
for combo in combinations(features, 2):
    print(combo)
```

```python
from itertools import accumulate

monthly_savings = [100, 150, 120, 180]
print(list(accumulate(monthly_savings)))
```

## 1.9 `functools`


`functools` contains tools for working with functions. Two very practical ones are `lru_cache` for memoization and `partial` for pre-filling function arguments.


```python
from functools import lru_cache

@lru_cache(maxsize=None)
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

print([fibonacci(n) for n in range(10)])
```

```python
from functools import partial

def power(base, exponent):
    return base ** exponent

square = partial(power, exponent=2)
cube = partial(power, exponent=3)

print(square(5))
print(cube(3))
```

## 1.10 `statistics`


`statistics` gives you core descriptive statistics directly from the standard library. It is useful for quick summaries without reaching for external libraries.


```python
from statistics import mean, median, stdev

values = [10, 12, 12, 13, 18]
print("Mean:", mean(values))
print("Median:", median(values))
print("Standard deviation:", round(stdev(values), 2))
```

```python
from statistics import multimode

responses = ["yes", "no", "yes", "maybe", "yes", "no"]
print("Most frequent values:", multimode(responses))
```

## 1.11 `re`


`re` lets you search and transform text with regular expressions. It is useful for extracting patterns such as emails, IDs, or formatted values.


```python
import re

text = "Contact alice@example.com or bob@python.org"
emails = re.findall(r"[\w.-]+@[\w.-]+", text)
print(emails)
```

```python
import re

log_line = "order=4821 status=paid total=199"
cleaned = re.sub(r"\d", "#", log_line)
print(cleaned)
```

## 1.12 `math`


`math` contains reliable mathematical functions and constants. It is useful for rounding, trigonometry, logarithms, and numeric calculations.


```python
import math

radius = 3
area = math.pi * radius ** 2
print("Circle area:", round(area, 2))
```

```python
import math

values = [1.2, 1.8, 2.1]
print("ceil:", [math.ceil(v) for v in values])
print("floor:", [math.floor(v) for v in values])
print("sqrt(81):", math.sqrt(81))
```

## 1.13 `random`


`random` helps you generate pseudo-random values for sampling, shuffling, and simulation. Setting a seed makes results reproducible for teaching and debugging.


```python
import random

random.seed(42)
print([random.randint(1, 10) for _ in range(5)])
```

```python
import random

random.seed(7)
items = ["numpy", "pandas", "matplotlib", "scikit-learn"]
print("Sample:", random.sample(items, 2))
random.shuffle(items)
print("Shuffled:", items)
```

## 1.14 `dataclasses`


`dataclasses` reduce boilerplate when creating small classes used mainly to store data. They automatically provide methods like `__init__` and a readable `__repr__`.


```python
from dataclasses import dataclass

@dataclass
class Student:
    name: str
    track: str
    level: str = "beginner"

student = Student(name="Alex", track="Python 3.12")
print(student)
```

```python
from dataclasses import dataclass

@dataclass(order=True)
class LeaderboardEntry:
    score: int
    name: str

entries = [LeaderboardEntry(88, "Bob"), LeaderboardEntry(95, "Alice"), LeaderboardEntry(91, "Cara")]
print(sorted(entries, reverse=True))
```

## 1.15 `enum`


`enum` lets you define a fixed set of named constants. It makes code clearer and safer when only a few valid states or categories are allowed.


```python
from enum import Enum

class Status(Enum):
    TODO = "todo"
    IN_PROGRESS = "in_progress"
    DONE = "done"

print(Status.TODO)
print(Status.DONE.value)
```

```python
from enum import IntEnum

class Priority(IntEnum):
    LOW = 1
    MEDIUM = 2
    HIGH = 3

print(Priority.HIGH > Priority.LOW)
print(list(Priority))
```
