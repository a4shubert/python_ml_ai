# 1. Basic Data Types and Core Values

This chapter refreshes the Python values that show up everywhere. For an educated adult returning to Python, the important questions are not only what each type stores, but whether it is mutable, ordered, hashable, exact, approximate, sequence-like, or structurally meaningful.

Much of Python fluency comes from choosing the right type for the job and understanding what that choice communicates to readers.

| Type | Ordered | Mutable | Hashable | Typical use |
|---|---|---|---|---|
| `int` / `float` / `complex` | n/a | No | Yes | Numeric computation |
| `bool` | n/a | No | Yes | Conditions and logic |
| `str` | Yes | No | Yes | Text |
| `bytes` | Yes | No | Yes | Binary payloads |
| `bytearray` | Yes | Yes | No | Mutable binary data |
| `list` | Yes | Yes | No | Evolving sequences |
| `tuple` | Yes | No | Usually | Fixed records |
| `set` / `frozenset` | No | `set`: Yes / `frozenset`: No | `frozenset`: Yes | Uniqueness and membership |
| `dict` | In insertion order | Yes | No | Named structure and lookup |
| `None` | n/a | No | Yes | Missing / absent value |

## 1.1 `int`, `float`, `complex`

`int` is exact for whole numbers, `float` is approximate for real-valued computation, and `complex` handles numbers with real and imaginary parts. The key refresher point is that the type is part of the meaning, not just storage.


```python
whole_number = 42
decimal_number = 3.75
complex_number = 2 + 3j

print(type(whole_number), whole_number)
print(type(decimal_number), decimal_number)
print(type(complex_number), complex_number)

# ----

x = 0.1 + 0.2
print("0.1 + 0.2 =", x)
print("Equals 0.3 exactly?", x == 0.3)
print("Magnitude of 2 + 3j:", abs(2 + 3j))
```

## 1.2 `bool` and truthiness

`bool` has only two values, but truthiness extends the idea to empty containers, zero values, and objects that define their own truth semantics. Real Python code relies on that constantly. In practice, the important refresher question is which operations the type supports, whether it is mutable, and what that implies for aliasing and correctness.


```python
print(True, False)
print(type(True))
print("True + True + False =", True + True + False)

# ----

samples = [0, 1, "", "python", [], [1], None]
print([(value, bool(value)) for value in samples])
```

## 1.4 `bytes` and `bytearray`

## 1.9 `None`, identity, and sentinels

```python
result = None
print(result is None)
print(type(result))

# ----

def find_even(numbers):
    for number in numbers:
        if number % 2 == 0:
            return number
    return None

print(find_even([1, 3, 5]) is None)
print(find_even([1, 4, 5]))
```

# 2. Expressions, Operators, and Control Flow

This chapter shifts from values to execution. The goal is to refresh how Python binds names, evaluates expressions, chooses branches, iterates, and encodes common control-flow patterns in a way that is explicit and readable.

A large amount of Python fluency comes from seeing that control flow is not just about syntax. It is about evaluation order, truthiness, short-circuiting, iterator consumption, and choosing a style that communicates intent clearly.

| Topic | Key question | Why it matters |
|---|---|---|
| Assignment | What does a name refer to? | Python binds names, not variables in the C-like sense |
| Operators | How is an expression evaluated? | Correctness depends on precedence, identity, and membership semantics |
| Branching | Which path runs and why? | Truthiness and matching shape readable decisions |
| Iteration | What is being consumed? | Loops often hide iterator semantics |
| Comprehensions | What are we transforming? | Expressive, but easy to over-compress |

## 2.1 Names, binding, and assignment

`=` in Python binds a name to an object; it does not copy values by default. This is a subtle but central idea, especially for mutable objects and function arguments.


```python
value = 10
alias = value
print(value, alias)

items = [1, 2]
other_name = items
other_name.append(3)
print(items, other_name)

# ----

x, y = 1, 2
x, y = y, x
print(x, y)

config = {"mode": "draft"}
backup = config
backup["mode"] = "final"
print(config)
```

## 2.2 Arithmetic, comparison, logical, membership, and identity operators

Operators in Python are expressive but semantically distinct. A refresher should emphasize the difference between value comparison, object identity, membership, and logical composition. The main teaching goal here is to connect syntax to control flow, because small operator and branching choices strongly affect how predictable code feels.


```python
a = 7
b = 3
print(a + b, a - b, a * b, a / b, a // b, a % b, a ** b)

# ----

name = "python"
other_name = "".join(["py", "thon"])
print(name == other_name)
print(name is other_name)
print("py" in name)
print((3 < 5) and (5 < 8))
```

## 2.3 Conditional expressions and short-circuiting

Conditional expressions let you express small decisions inline. Short-circuiting means logical operators stop once the result is already determined, which affects both efficiency and side effects. The main teaching goal here is to connect syntax to control flow, because small operator and branching choices strongly affect how predictable code feels.


```python
score = 72
status = "pass" if score >= 50 else "fail"
print(status)

# ----

def expensive_check():
    print("expensive check ran")
    return True

print(False and expensive_check())
print(True or expensive_check())
```

## 2.4 `if` / `elif` / `else`

Branching is where truthiness meets explicit decision-making. The main design question is not how to branch, but how to make branch logic readable and unsurprising. The main teaching goal here is to connect syntax to control flow, because small operator and branching choices strongly affect how predictable code feels.


```python
temperature = 18
if temperature < 10:
    label = "cold"
elif temperature < 20:
    label = "mild"
else:
    label = "warm"
print(label)

# ----

user_input = ""
if user_input:
    print("received:", user_input)
else:
    print("input was empty")
```

## 2.5 `for` and `while`

`for` expresses iteration over an iterable, while `while` expresses repetition under a condition. In Python, `for` is often preferred because it is safer and more declarative. The main teaching goal here is to connect syntax to control flow, because small operator and branching choices strongly affect how predictable code feels.


```python
for index, letter in enumerate("abc", start=1):
    print(index, letter)

# ----

countdown = 3
while countdown > 0:
    print(countdown)
    countdown -= 1
print("done")
```

## 2.6 `break`, `continue`, `pass`

These statements alter loop behavior directly. The main nuance is to use them where they simplify control flow, not where they make the path through a loop harder to reconstruct mentally. The main teaching goal here is to connect syntax to control flow, because small operator and branching choices strongly affect how predictable code feels.


```python
for number in range(1, 6):
    if number == 4:
        break
    print(number)

# ----

for number in range(1, 6):
    if number % 2 == 0:
        continue
    print("odd:", number)

if True:
    pass
```

## 2.7 `match` / `case`

`match` adds structural pattern matching to Python. It is most useful when code is branching on shape or category rather than just boolean conditions. The main teaching goal here is to connect syntax to control flow, because small operator and branching choices strongly affect how predictable code feels.


```python
command = ("move", 10, 20)

match command:
    case ("move", x, y):
        print(f"moving to {x}, {y}")
    case ("quit",):
        print("quitting")
    case _:
        print("unknown command")

# ----

status_code = 404
match status_code:
    case 200:
        print("ok")
    case 400 | 404:
        print("client-side issue")
    case _:
        print("other")
```

## 2.8 Comprehensions and generator expressions

Comprehensions are concise transformation tools; generator expressions provide lazy evaluation. The refresh point is that these are not just shorter loops, they encode a particular style of thinking. The main teaching goal here is to connect syntax to control flow, because small operator and branching choices strongly affect how predictable code feels.


```python
numbers = range(8)
squares_of_even = [n * n for n in numbers if n % 2 == 0]
print(squares_of_even)

# ----

numbers = range(8)
lazy_squares = (n * n for n in numbers if n % 2 == 0)
print(lazy_squares)
print(list(lazy_squares))
```

## 2.9 Common control-flow idioms

Experienced Python code often uses compact patterns such as `any`, `all`, `enumerate`, sentinel checks, and guard clauses. These idioms compress intent without obscuring the logic. The main teaching goal here is to connect syntax to control flow, because small operator and branching choices strongly affect how predictable code feels.


```python
values = [2, 4, 6, 8]
print(any(v > 5 for v in values))
print(all(v % 2 == 0 for v in values))

# ----

records = ["alpha", "beta", "gamma"]
for index, record in enumerate(records, start=1):
    print(f"{index}: {record}")
```

## 4.4 Iteration protocol

## 9.1 Class definition and instance state

A class defines the behavior shared by instances; each instance holds its own state. The key design question is whether the state and operations naturally belong together.


```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

alice = Student("Alice", 95)
bob = Student("Bob", 84)
print(alice.name, alice.score)
print(bob.name, bob.score)

# ----

class Task:
    def __init__(self, title, done=False):
        self.title = title
        self.done = done

    def complete(self):
        self.done = True

task = Task("Write notebook")
task.complete()
print(task.title, task.done)
```

## 9.2 Class variables vs instance variables

Class variables are shared across instances; instance variables belong to each object separately. Confusing the two leads to subtle state bugs and misleading APIs. The educational goal is not merely to define classes, but to see when object-oriented structure clarifies a problem and when it adds noise.


```python
class BankAccount:
    interest_rate = 0.02

    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

a = BankAccount("Alice", 100)
b = BankAccount("Bob", 200)
print(a.interest_rate, b.interest_rate)

# ----

a.balance += 50
BankAccount.interest_rate = 0.03
print(a.balance, b.balance)
print(a.interest_rate, b.interest_rate)
```

## 9.3 Methods: instance, class, static

Python supports instance methods, class methods, and static methods, each for a different purpose. Use them intentionally rather than decoratively. The educational goal is not merely to define classes, but to see when object-oriented structure clarifies a problem and when it adds noise.


```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    def to_fahrenheit(self):
        return self.celsius * 9 / 5 + 32

print(Temperature(20).to_fahrenheit())

# ----

class User:
    def __init__(self, name):
        self.name = name

    @classmethod
    def guest(cls):
        return cls("guest")

    @staticmethod
    def is_valid_name(name):
        return bool(name.strip())

print(User.guest().name)
print(User.is_valid_name(" Alice "))
```

## 9.4 Inheritance and composition

Inheritance models “is-a” specialization, while composition builds larger behavior from smaller parts. In Python, composition is often the more robust default.
```python
class Animal:
    def speak(self):
        return "..."

class Dog(Animal):
    def speak(self):
        return "woof"

print(Dog().speak())

# ----

class Engine:
    def start(self):
        return "engine started"

class Car:
    def __init__(self):
        self.engine = Engine()

    def start(self):
        return self.engine.start()

print(Car().start())
```

## 9.8 Protocol-oriented vs inheritance-heavy design

In Python, objects often work because they support the right operations, not because they inherit from a particular base class. This usually leads to simpler and more flexible designs. ## 10.8 Readability tradeoffs

Functional style is helpful only while it improves clarity.
```python
ugly = list(filter(lambda x: x % 2 == 0, map(lambda x: x * 2, range(6))))
clear = [x * 2 for x in range(6) if x % 2 == 0]

print("compressed style:", ugly)
print("clearer style:", clear)

# ----

def transformed(values):
    result = []
    for value in values:
        if value % 2 == 0:
            result.append(value * 2)
    return result

print(transformed(range(6)))
```

## 13.7 Async iteration and async context managers

Async code extends the usual iteration and context-manager ideas into awaitable forms. These patterns matter in real async libraries even if the syntax appears less often in small scripts.
```python
import asyncio

class AsyncCounter:
    def __init__(self, limit):
        self.current = 0
        self.limit = limit

    def __aiter__(self):
        return self

    async def __anext__(self):
        if self.current >= self.limit:
            raise StopAsyncIteration
        self.current += 1
        await asyncio.sleep(0.001)
        return self.current

async def consume():
    values = []
    async for item in AsyncCounter(3):
        values.append(item)
    print(values)

await consume()

# ----

import asyncio

class AsyncResource:
    async def __aenter__(self):
        print("async open")
        return self

    async def __aexit__(self, exc_type, exc, tb):
        print("async close")

async def use_resource():
    async with AsyncResource():
        print("inside async context")

await use_resource()
```

## 14.3 Decorators

Decorators wrap functions or classes with reusable behavior. They are powerful because they let cross-cutting concerns such as logging, timing, caching, or validation be attached declaratively. These features are powerful because they expose how Python can inspect and modify behavior at runtime.


```python
def emphasize(func):
    def wrapper(*args, **kwargs):
        return f"**{func(*args, **kwargs)}**"
    return wrapper

@emphasize
def greet(name):
    return f"hello {name}"

print(greet("alice"))

# ----

def log_call(func):
    def wrapper(*args, **kwargs):
        print(f"calling {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

@log_call
def square(x):
    return x * x

print(square(5))
```

```python
class UpperCase:
    def __get__(self, instance, owner):
        return instance._name.upper()

    def __set__(self, instance, value):
        instance._name = value

class Person:
    name = UpperCase()

    def __init__(self, name):
        self.name = name

p = Person("alice")
print(p.name)

# ----

class Number:
    def __set_name__(self, owner, name):
        self.private_name = "_" + name

    def __get__(self, instance, owner):
        return getattr(instance, self.private_name)

    def __set__(self, instance, value):
        setattr(instance, self.private_name, int(value))

class Counter:
    count = Number()

    def __init__(self, count):
        self.count = count

print(Counter(3.8).count)
```

## 16.5 Refactoring procedural code into better abstractions

Procedural code is often the right starting point, but it can become hard to maintain when responsibilities pile up. Refactoring means extracting structure only when it clarifies behavior.
```python
def report_line(name, score):
    return f"{name}: {score}"

def report_block(records):
    return "\n".join(report_line(name, score) for name, score in records)

print(report_block([("Alice", 95), ("Bob", 84)]))

# ----

records = [("Alice", 95), ("Bob", 84)]
formatted = [f"{name}: {score}" for name, score in records]
print(formatted)
```
