## 6.11.1 Regression metrics

### Why It Matters
MAE, RMSE, and R-squared emphasize different error properties. MAE is often easier to explain, while RMSE punishes large misses more heavily.


```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_true = np.array([52, 60, 66, 79, 85], dtype=float)
y_pred = np.array([50, 63, 64, 76, 88], dtype=float)
{
    "mae": round(float(mean_absolute_error(y_true, y_pred)), 3),
    "rmse": round(float(mean_squared_error(y_true, y_pred) ** 0.5), 3),
    "r2": round(float(r2_score(y_true, y_pred)), 3),
}
```

## 6.11.2 Classification metrics

### Why It Matters
Accuracy, precision, recall, F1, and ROC-AUC each focus on different operational tradeoffs.


```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_true = np.array([1,1,1,0,0,0,1,0])
y_pred = np.array([1,1,0,0,1,0,1,0])
{
    "accuracy": round(float(accuracy_score(y_true, y_pred)), 3),
    "precision": round(float(precision_score(y_true, y_pred)), 3),
    "recall": round(float(recall_score(y_true, y_pred)), 3),
    "f1": round(float(f1_score(y_true, y_pred)), 3),
}
```

## 6.11.4 Threshold tuning

### Why It Matters
Probability models let you trade recall for precision or vice versa by moving the decision threshold.


```python
import numpy as np
from sklearn.metrics import precision_score, recall_score

y_true = np.array([1,1,1,0,0,0,1,0])
probs = np.array([0.95,0.72,0.48,0.61,0.43,0.20,0.80,0.31])
summary = {}
for threshold in [0.4, 0.6, 0.8]:
    preds = (probs >= threshold).astype(int)
    summary[threshold] = {
        "precision": round(float(precision_score(y_true, preds)), 3),
        "recall": round(float(recall_score(y_true, preds)), 3),
    }
summary
```

## 6.11.6 Metric alignment

### Why It Matters
The best metric is the one that approximates the decision cost of real mistakes.


```python
import numpy as np
from sklearn.metrics import precision_score, recall_score

y_true = np.array([1,1,1,0,0,0,1,0,1,0])
probs = np.array([0.94,0.78,0.66,0.61,0.42,0.20,0.73,0.31,0.55,0.15])
summary = {}
for threshold in [0.4, 0.7]:
    preds = (probs >= threshold).astype(int)
    summary[threshold] = {
        "precision": round(float(precision_score(y_true, preds)), 3),
        "recall": round(float(recall_score(y_true, preds)), 3),
    }
summary
```
