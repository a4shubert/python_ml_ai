## 6.14.1 k-means intuition

### Why It Matters
k-means alternates between assigning points to centroids and updating centroid locations. It is fast but assumes roughly spherical clusters.


```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=200, centers=4, random_state=71)
model = KMeans(n_clusters=4, n_init=20, random_state=71).fit(X)
model.cluster_centers_.round(2)
```

## 6.14.2 Choosing k

### Why It Matters
There is no universal correct `k`; elbow plots, silhouette scores, and domain needs all contribute to the decision.


```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=220, centers=4, random_state=72)
results = {}
for k in [2, 3, 4, 5]:
    labels = KMeans(n_clusters=k, n_init=20, random_state=72).fit_predict(X)
    results[k] = round(float(silhouette_score(X, labels)), 3)
results
```

## 6.14.3 Initialization sensitivity

### Why It Matters
Different starts can yield different local optima, so repeated restarts matter.


```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=180, centers=3, random_state=73)
model_a = KMeans(n_clusters=3, n_init=1, random_state=1).fit(X)
model_b = KMeans(n_clusters=3, n_init=1, random_state=2).fit(X)
{"inertia_a": round(float(model_a.inertia_), 2), "inertia_b": round(float(model_b.inertia_), 2)}
```

## 6.14.4 Hierarchical clustering

### Why It Matters
Hierarchical methods build a nested tree of merges or splits rather than a single flat partition.


```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=150, centers=3, random_state=74)
labels = AgglomerativeClustering(n_clusters=3).fit_predict(X)
list(labels[:10])
```

## 6.14.5 Linkage choices

### Why It Matters
Single, complete, and average linkage create different shapes of cluster trees because they define inter-cluster distance differently.


```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=180, centers=3, random_state=75)
labels_complete = AgglomerativeClustering(n_clusters=3, linkage="complete").fit_predict(X)
labels_average = AgglomerativeClustering(n_clusters=3, linkage="average").fit_predict(X)
{"complete_unique": int(len(set(labels_complete))), "average_unique": int(len(set(labels_average)))}
```

## 6.14.6 Interpretation

### Why It Matters
These methods are best used as exploratory summarizers unless the domain gives a strong reason to treat the clusters as operational groups.


```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=160, centers=3, random_state=76)
labels = KMeans(n_clusters=3, n_init=20, random_state=76).fit_predict(X)
{int(label): int((labels == label).sum()) for label in sorted(set(labels))}
```
