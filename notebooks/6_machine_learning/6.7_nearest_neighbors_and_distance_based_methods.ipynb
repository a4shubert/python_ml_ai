## 6.7.1 Local decision rules

### Why It Matters
k-NN predicts from nearby observed points. This makes it flexible but sensitive to local noise and sample density.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier

X, y = make_classification(n_samples=140, n_features=2, n_redundant=0, n_informative=2, random_state=23)
model = KNeighborsClassifier(n_neighbors=5).fit(X, y)
model.predict(X[:5]).tolist()
```

## 6.7.3 Scaling sensitivity

### Why It Matters
Unscaled features can dominate the neighbor search and distort results.


```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

X = np.array([[10, 0.1], [12, 0.2], [50, 0.9], [55, 1.0]])
y = np.array([0, 0, 1, 1])
scaled = StandardScaler().fit_transform(X)
model = KNeighborsClassifier(n_neighbors=1).fit(scaled, y)
model.predict(scaled).tolist()
```

## 6.7.4 Choosing k

### Why It Matters
Small `k` tracks local detail but can overfit; larger `k` smooths predictions and reduces variance.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

X, y = make_classification(n_samples=240, n_features=6, random_state=24)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=24)
results = {}
for k in [1, 5, 15]:
    model = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    results[k] = round(float(accuracy_score(y_test, model.predict(X_test))), 3)
results
```

## 6.7.5 Curse of dimensionality

### Why It Matters
As dimension grows, distances become less informative and neighborhoods become less distinct.


```python
import numpy as np

rng = np.random.default_rng(17)
ratios = {}
for dim in [2, 50]:
    points = rng.normal(size=(300, dim))
    query = points[0]
    dists = np.linalg.norm(points[1:] - query, axis=1)
    ratios[dim] = round(float(dists.min() / dists.max()), 3)
ratios
```

## 6.7.6 When to use

### Why It Matters
k-NN can work well on moderate-dimensional problems with meaningful local structure and enough labeled data.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

results = {}
for features, random_state in [(2, 18), (50, 19)]:
    X, y = make_classification(n_samples=300, n_features=features, n_informative=max(2, features // 5), n_redundant=0, random_state=random_state)
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=random_state)
    model = KNeighborsClassifier(n_neighbors=7).fit(X_train, y_train)
    results[f"{features}_features_accuracy"] = round(float(accuracy_score(y_test, model.predict(X_test))), 3)
results
```
