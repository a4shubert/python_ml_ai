## 6.10.1 Margin intuition

### Why It Matters
SVMs look for a boundary that separates classes with the widest possible margin. Wider margins often generalize better.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.svm import SVC

X, y = make_classification(n_samples=150, n_features=2, n_redundant=0, n_informative=2, class_sep=1.5, random_state=39)
model = SVC(kernel="linear", C=1.0).fit(X, y)
round(float(1 / np.linalg.norm(model.coef_)), 3)
```

## 6.10.2 Soft margins

### Why It Matters
Allowing some violations prevents the model from overfitting noisy or overlapping data.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.svm import SVC

X, y = make_classification(n_samples=160, n_features=2, n_redundant=0, n_informative=2, class_sep=1.0, random_state=40)
soft = SVC(kernel="linear", C=0.1).fit(X, y)
hard = SVC(kernel="linear", C=10.0).fit(X, y)
{"support_vectors_soft": int(soft.n_support_.sum()), "support_vectors_hard": int(hard.n_support_.sum())}
```

## 6.10.3 Kernel idea

### Why It Matters
Kernels let SVMs represent nonlinear boundaries without explicitly expanding all features by hand.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.svm import SVC

X, y = make_classification(n_samples=180, n_features=2, n_redundant=0, n_informative=2, class_sep=0.9, random_state=41)
linear = SVC(kernel="linear").fit(X, y)
rbf = SVC(kernel="rbf", gamma="scale").fit(X, y)
{"linear_train_score": round(float(linear.score(X, y)), 3), "rbf_train_score": round(float(rbf.score(X, y)), 3)}
```

## 6.10.4 Scaling and tuning

### Why It Matters
SVMs are sensitive to feature scale and hyperparameters such as `C` and kernel parameters.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

X, y = make_classification(n_samples=180, n_features=4, random_state=42)
scaled = StandardScaler().fit_transform(X)
model = SVC(kernel="rbf", C=1.0, gamma="scale").fit(scaled, y)
round(float(model.score(scaled, y)), 3)
```

## 6.10.5 Strengths and weaknesses

### Why It Matters
SVMs can be powerful, but they are harder to interpret and slower to tune than linear models.


```python
import numpy as np
from sklearn.datasets import make_moons
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

X, y = make_moons(n_samples=260, noise=0.18, random_state=24)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=24)
logit = LogisticRegression(max_iter=1000).fit(X_train, y_train)
svm = SVC(kernel="rbf", C=1.0, gamma="scale").fit(X_train, y_train)
{
    "logistic_test_accuracy": round(float(accuracy_score(y_test, logit.predict(X_test))), 3),
    "rbf_svm_test_accuracy": round(float(accuracy_score(y_test, svm.predict(X_test))), 3),
}
```

## 6.10.6 Use when needed

### Why It Matters
SVMs are worth considering when linear boundaries underfit and tabular datasets are not enormous.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

X, y = make_classification(n_samples=320, n_features=8, n_informative=5, random_state=25)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=25)
logit = LogisticRegression(max_iter=1000).fit(X_train, y_train)
svm = SVC(kernel="rbf", C=1.0, gamma="scale").fit(X_train, y_train)
{
    "logistic_accuracy": round(float(accuracy_score(y_test, logit.predict(X_test))), 3),
    "svm_accuracy": round(float(accuracy_score(y_test, svm.predict(X_test))), 3),
}
```
