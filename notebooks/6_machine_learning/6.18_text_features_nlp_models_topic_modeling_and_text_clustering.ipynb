## 6.18.1 Bag-of-words and TF-IDF

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

docs = ["customer retention improved", "retention depends on onboarding"]
counts = CountVectorizer().fit_transform(docs)
tfidf = TfidfVectorizer().fit_transform(docs)
{"count_shape": counts.shape, "tfidf_shape": tfidf.shape}
```

## 6.18.2 n-grams

## 6.18.3 Text classification

### Why It Matters
Linear models such as logistic regression often work well on sparse text features because the representation is already high-dimensional.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

texts = ["strong revenue growth", "customer renewal discussion", "match report victory", "team defense tactics"]
labels = ["business", "business", "sports", "sports"]
X = TfidfVectorizer().fit_transform(texts)
model = LogisticRegression(max_iter=1000).fit(X, labels)
model.predict(X).tolist()
```

## 6.18.4 Topic modeling intuition

### Why It Matters
Topic models or matrix factorization methods summarize recurring term patterns across documents.


```python
from sklearn.decomposition import NMF
from sklearn.feature_extraction.text import TfidfVectorizer

texts = ["revenue growth and margin", "customer renewal and pricing", "match report and tactics", "defense and midfield control"]
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(texts)
model = NMF(n_components=2, init='nndsvda', random_state=101, max_iter=400).fit(X)
terms = vectorizer.get_feature_names_out()
[[terms[i] for i in comp.argsort()[-3:][::-1]] for comp in model.components_]
```

## 6.18.5 Text clustering

### Why It Matters
Vectorized documents can be grouped with clustering algorithms to reveal rough thematic structure.


```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer

texts = ["revenue growth and margin", "customer renewal and pricing", "match report and tactics", "defense and midfield control"]
X = TfidfVectorizer(stop_words='english').fit_transform(texts)
labels = KMeans(n_clusters=2, n_init=20, random_state=102).fit_predict(X)
labels.tolist()
```

## 6.18.6 Interpretation limits

### Why It Matters
Topics and clusters are summaries, not objective semantic truths; they require human interpretation.


```python
from sklearn.decomposition import NMF
from sklearn.feature_extraction.text import TfidfVectorizer

texts = ["revenue growth and margin", "customer renewal and pricing", "match report and tactics", "defense and midfield control"]
X = TfidfVectorizer(stop_words='english').fit_transform(texts)
model = NMF(n_components=2, init='nndsvda', random_state=103, max_iter=400).fit(X)
{"topic_count": model.components_.shape[0], "needs_human_labeling": True}
```
