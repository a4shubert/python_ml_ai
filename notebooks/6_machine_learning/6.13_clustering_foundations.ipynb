## 6.13.1 What clustering can and cannot do

### Why It Matters
Clustering can summarize structure, segment data, or support exploration. It does not automatically discover “real” categories.


```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import adjusted_rand_score

X, true_labels = make_blobs(n_samples=200, centers=3, cluster_std=1.1, random_state=31)
pred_labels = KMeans(n_clusters=3, n_init=20, random_state=31).fit_predict(X)
round(float(adjusted_rand_score(true_labels, pred_labels)), 3)
```

## 6.13.2 Similarity choices

### Why It Matters
Distance metrics and feature scales determine which points look close. Clusters can change dramatically when those choices change.


```python
import numpy as np
from scipy.spatial.distance import cdist

points = np.array([[0,0],[1,1],[3,1]], dtype=float)
cdist(points[:1], points[1:]).round(3)
```

## 6.13.3 Scaling and representation

### Why It Matters
Standardization is often essential because high-variance features otherwise dominate the grouping.


```python
import numpy as np
from sklearn.preprocessing import StandardScaler

X = np.array([[1000, 1], [1200, 2], [1500, 1], [1800, 3]], dtype=float)
StandardScaler().fit_transform(X).round(3)
```

## 6.13.4 Clusterability

### Why It Matters
Not every dataset truly contains useful clusters. Sometimes the strongest result is learning that clean segmentation is weak.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.decomposition import PCA

X, _ = make_classification(n_samples=220, n_features=10, n_informative=4, random_state=61)
explained = PCA(n_components=2).fit(X).explained_variance_ratio_
explained.round(3).tolist()
```

## 6.13.5 Validation caution

### Why It Matters
Internal scores help compare clusterings, but they cannot replace domain interpretation and external usefulness.


```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=180, centers=4, random_state=62)
labels = KMeans(n_clusters=4, n_init=20, random_state=62).fit_predict(X)
round(float(silhouette_score(X, labels)), 3)
```

## 6.13.6 Use cases

### Why It Matters
Segmentation, anomaly investigation, document grouping, and exploratory pattern-finding are common clustering applications.


```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=240, centers=[[0,0],[4,4],[0,5]], cluster_std=[0.5,0.8,0.6], random_state=32)
labels = KMeans(n_clusters=3, n_init=20, random_state=32).fit_predict(X)
{int(label): int((labels == label).sum()) for label in sorted(set(labels))}
```
