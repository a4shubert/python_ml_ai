## 6.17.2 Tokenization

### Why It Matters
Tokenization decides what counts as a unit: words, subwords, or n-grams. In classical NLP, word-level tokens are the common starting point.


```python
text = "Revenue growth depends on customer retention."
tokens = text.lower().replace('.', '').split()
[(i, tok) for i, tok in enumerate(tokens)]
```

## 6.17.3 Normalization

### Why It Matters
Lowercasing, punctuation stripping, and simple normalization reduce superficial variation that can fragment the vocabulary.


```python
raw = ["  Revenue!", "Growth,", "retention  "]
normalized = [token.strip().lower().replace('!', '').replace(',', '') for token in raw]
normalized
```

## 6.17.4 Stopwords and pruning

### Why It Matters
Very common words may add little predictive value and can be filtered, though blanket removal is not always wise.


```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

doc = "the retention of the customer depends on the onboarding flow".split()
[token for token in doc if token not in ENGLISH_STOP_WORDS]
```

## 6.17.5 Stemming and lemmatization intuition

### Why It Matters
Reducing related word forms can shrink the vocabulary and expose semantic similarity, but aggressive stemming can also destroy nuance.


```python
words = ["connect", "connected", "connection", "connecting"]
stem_like = [w[:7] for w in words]
list(zip(words, stem_like))
```

## 6.17.6 Sparse representations

### Why It Matters
Classical NLP usually produces sparse matrices with many zeros. That structure is normal and often efficient.


```python
from sklearn.feature_extraction.text import CountVectorizer

docs = ["customer retention improved", "retention depends on onboarding"]
X = CountVectorizer().fit_transform(docs)
{"shape": X.shape, "nonzero_entries": int(X.nnz)}
```
