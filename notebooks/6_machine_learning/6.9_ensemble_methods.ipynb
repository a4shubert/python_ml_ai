## 6.9.1 Bagging intuition

### Why It Matters
Bagging reduces variance by averaging many models fit on resampled data. Random forests are the classic example.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier

X, y = make_classification(n_samples=220, n_features=6, random_state=33)
model = RandomForestClassifier(n_estimators=120, max_depth=5, random_state=33).fit(X, y)
round(float(model.score(X, y)), 3)
```

## 6.9.2 Random forests

### Why It Matters
Random forests decorrelate trees by sampling features as well as rows. This produces robust performance on many tabular problems.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier

X, y = make_classification(n_samples=200, n_features=5, random_state=34)
forest = RandomForestClassifier(n_estimators=100, random_state=34).fit(X, y)
dict(zip([f"x{i}" for i in range(1,6)], forest.feature_importances_.round(3)))
```

## 6.9.3 Boosting intuition

### Why It Matters
Boosting fits models sequentially, focusing new learners on the mistakes of previous ones.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.ensemble import GradientBoostingClassifier

X, y = make_classification(n_samples=220, n_features=6, random_state=35)
boost = GradientBoostingClassifier(random_state=35).fit(X, y)
round(float(boost.score(X, y)), 3)
```

## 6.9.4 Bias and variance tradeoffs

### Why It Matters
Bagging mainly reduces variance, while boosting often reduces bias and variance at the cost of more tuning sensitivity.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

X, y = make_classification(n_samples=320, n_features=10, n_informative=6, random_state=22)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=22)
rf = RandomForestClassifier(n_estimators=200, max_depth=6, random_state=22).fit(X_train, y_train)
gb = GradientBoostingClassifier(random_state=22).fit(X_train, y_train)
{
    "random_forest_train": round(float(accuracy_score(y_train, rf.predict(X_train))), 3),
    "random_forest_test": round(float(accuracy_score(y_test, rf.predict(X_test))), 3),
    "gradient_boosting_train": round(float(accuracy_score(y_train, gb.predict(X_train))), 3),
    "gradient_boosting_test": round(float(accuracy_score(y_test, gb.predict(X_test))), 3),
}
```

## 6.9.5 Feature importance caveats

### Why It Matters
Importance summaries can help, but they can also mislead when predictors are correlated or when importance is defined inconsistently.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier

X, y = make_classification(n_samples=260, n_features=8, n_informative=4, random_state=36)
forest = RandomForestClassifier(n_estimators=150, random_state=36).fit(X, y)
important = sorted(zip(range(X.shape[1]), forest.feature_importances_), key=lambda x: x[1], reverse=True)[:3]
[(f"x{i+1}", round(float(v), 3)) for i, v in important]
```

## 6.9.6 Operational tradeoffs

### Why It Matters
Ensembles often win on raw accuracy, but they may be harder to explain and maintain than strong linear baselines.


```python
import time
import numpy as np
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

X, y = make_classification(n_samples=500, n_features=12, n_informative=6, random_state=23)
start = time.perf_counter(); LogisticRegression(max_iter=1000).fit(X, y); logit_time = time.perf_counter() - start
start = time.perf_counter(); RandomForestClassifier(n_estimators=200, random_state=23).fit(X, y); rf_time = time.perf_counter() - start
{"logistic_fit_seconds": round(logit_time, 3), "random_forest_fit_seconds": round(rf_time, 3)}
```
