## 6.15.1 DBSCAN intuition

### Why It Matters
DBSCAN finds dense regions and labels sparse points as noise. It is valuable when outliers or irregular cluster shapes matter.


```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, _ = make_moons(n_samples=220, noise=0.08, random_state=81)
labels = DBSCAN(eps=0.22, min_samples=5).fit_predict(X)
{int(label): int((labels == label).sum()) for label in sorted(set(labels))}
```

## 6.15.2 Neighborhood parameters

### Why It Matters
DBSCAN depends heavily on `eps` and `min_samples`, which define what density means in practice.


```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, _ = make_moons(n_samples=220, noise=0.08, random_state=82)
small_eps = DBSCAN(eps=0.15, min_samples=5).fit_predict(X)
large_eps = DBSCAN(eps=0.3, min_samples=5).fit_predict(X)
{"clusters_small_eps": len(set(small_eps)) - (1 if -1 in small_eps else 0), "clusters_large_eps": len(set(large_eps)) - (1 if -1 in large_eps else 0)}
```

## 6.15.3 Gaussian mixtures

### Why It Matters
Gaussian mixtures model data as draws from overlapping probabilistic components rather than hard partitions.


```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture

X, _ = make_blobs(n_samples=200, centers=3, cluster_std=1.4, random_state=83)
model = GaussianMixture(n_components=3, random_state=83).fit(X)
model.means_.round(2)
```

## 6.15.4 Soft assignments

### Why It Matters
A point can plausibly belong partly to multiple components. That soft membership is useful in ambiguous regions.


```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture

X, _ = make_blobs(n_samples=150, centers=2, cluster_std=1.8, random_state=84)
model = GaussianMixture(n_components=2, random_state=84).fit(X)
model.predict_proba(X[:5]).round(3)
```

## 6.15.5 Shape assumptions

### Why It Matters
Mixture models assume component families, while DBSCAN assumes density-connected clusters. Both can fail if those assumptions mismatch the geometry.


```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.mixture import GaussianMixture

X, _ = make_moons(n_samples=220, noise=0.1, random_state=33)
db_labels = DBSCAN(eps=0.2, min_samples=5).fit_predict(X)
gmm_labels = GaussianMixture(n_components=2, random_state=33).fit_predict(X)
{"dbscan_clusters": len(set(db_labels)) - (1 if -1 in db_labels else 0), "gmm_clusters": len(set(gmm_labels))}
```

## 6.15.6 Noise handling

### Why It Matters
Explicit noise labeling is sometimes more realistic than forcing every point into a cluster.


```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, _ = make_moons(n_samples=200, noise=0.1, random_state=85)
labels = DBSCAN(eps=0.2, min_samples=5).fit_predict(X)
{"noise_points": int((labels == -1).sum())}
```
