## 6.6.1 Probability outputs

### Why It Matters
Logistic regression estimates class probability rather than only a hard label. That is useful for ranking and threshold tuning.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

X, y = make_classification(n_samples=160, n_features=4, random_state=17)
model = LogisticRegression(max_iter=1000).fit(X, y)
model.predict_proba(X[:5]).round(3)
```

## 6.6.2 Decision thresholds

### Why It Matters
A 0.5 threshold is a default, not a law. Different error costs often justify a different cutoff.


```python
import numpy as np
from sklearn.metrics import precision_score, recall_score

y_true = np.array([1, 1, 1, 0, 0, 0, 1, 0])
probs = np.array([0.92, 0.77, 0.55, 0.51, 0.45, 0.30, 0.62, 0.20])
preds_05 = (probs >= 0.5).astype(int)
preds_07 = (probs >= 0.7).astype(int)
{
    "precision_at_0.5": round(float(precision_score(y_true, preds_05)), 3),
    "recall_at_0.5": round(float(recall_score(y_true, preds_05)), 3),
    "precision_at_0.7": round(float(precision_score(y_true, preds_07)), 3),
    "recall_at_0.7": round(float(recall_score(y_true, preds_07)), 3),
}
```

## 6.6.3 Class imbalance

### Why It Matters
When one class is rare, accuracy can hide poor minority detection. Thresholds and class weights become more important.


```python
import numpy as np
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, recall_score

X = np.arange(100).reshape(-1, 1)
y = np.array([0] * 92 + [1] * 8)
model = DummyClassifier(strategy="most_frequent").fit(X, y)
preds = model.predict(X)
{"accuracy": round(float(accuracy_score(y, preds)), 3), "minority_recall": round(float(recall_score(y, preds)), 3)}
```

## 6.6.4 Linear decision boundaries

### Why It Matters
Logistic regression works best when the relationship is roughly linear on the log-odds scale or when features encode the needed nonlinearity.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

X, y = make_classification(n_samples=180, n_features=2, n_redundant=0, n_informative=2, class_sep=1.5, random_state=19)
model = LogisticRegression(max_iter=1000).fit(X, y)
{"coef_1": round(float(model.coef_[0][0]), 3), "coef_2": round(float(model.coef_[0][1]), 3)}
```

## 6.6.5 Regularization

### Why It Matters
Shrinkage improves stability and prevents overreaction to noisy patterns.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

X, y = make_classification(n_samples=180, n_features=8, random_state=20)
model = LogisticRegression(max_iter=1000, C=0.3).fit(X, y)
round(float(np.linalg.norm(model.coef_)), 3)
```
