## 6.16.1 PCA intuition

### Why It Matters
PCA rotates the data into orthogonal directions that capture as much variance as possible with fewer components.


```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import make_classification

X, _ = make_classification(n_samples=220, n_features=8, n_informative=4, random_state=91)
pca = PCA(n_components=2).fit(X)
pca.components_.round(3)
```

## 6.16.2 Explained variance

### Why It Matters
The explained variance ratio helps judge how much information remains after compression.


```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import make_classification

X, _ = make_classification(n_samples=220, n_features=10, n_informative=4, random_state=92)
PCA(n_components=3).fit(X).explained_variance_ratio_.round(3).tolist()
```

## 6.16.3 Visualization use

### Why It Matters
Reducing to two components can reveal rough groupings or outliers, but a pretty projection is not a proof of clean separability.


```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=180, centers=3, n_features=5, random_state=93)
coords = PCA(n_components=2).fit_transform(X)
coords[:5].round(2).tolist()
```

## 6.16.4 Anomaly detection framing

### Why It Matters
Anomalies are context-dependent. A point can be rare statistically but normal operationally, or the reverse.


```python
import numpy as np
from sklearn.ensemble import IsolationForest

rng = np.random.default_rng(94)
X = rng.normal(size=(180, 2))
X[:5] += 6
model = IsolationForest(contamination=0.05, random_state=94).fit(X)
model.predict(X[:10]).tolist()
```

## 6.16.5 Isolation-style logic

### Why It Matters
Isolation methods flag points that are easier to separate from the bulk of the data than typical observations.


```python
import numpy as np
from sklearn.ensemble import IsolationForest

rng = np.random.default_rng(95)
X = rng.normal(size=(150, 2))
X[:4] += 5
scores = IsolationForest(contamination=0.05, random_state=95).fit(X).decision_function(X[:10])
scores.round(3).tolist()
```

## 6.16.6 Operational caution

### Why It Matters
Anomaly systems need feedback loops because many rare points are harmless and many harmful points are not that rare.


```python
import numpy as np
from sklearn.ensemble import IsolationForest

rng = np.random.default_rng(34)
X = rng.normal(size=(220, 2))
X[:6] += 5
model = IsolationForest(contamination=0.03, random_state=34).fit(X)
labels = model.predict(X)
{"flagged_as_anomalies": int((labels == -1).sum()), "flagged_as_normal": int((labels == 1).sum())}
```
