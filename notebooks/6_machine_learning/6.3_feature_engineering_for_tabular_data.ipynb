## 6.3.1 Numeric scaling

### Why It Matters
Distance-based and margin-based models are sensitive to scale. Trees usually are not, so preprocessing should match the model family.


```python
import numpy as np
from sklearn.preprocessing import StandardScaler

X = np.array([[150, 1.2], [175, 1.5], [210, 1.8]], dtype=float)
scaled = StandardScaler().fit_transform(X)
scaled.round(3)
```

## 6.3.2 Categorical encoding

### Why It Matters
One-hot encoding expands categories into indicator columns so linear and distance-based models can use them.


```python
import pandas as pd

categories = pd.DataFrame({"region": ["north", "south", "east", "north"]})
pd.get_dummies(categories, columns=["region"], dtype=int)
```

## 6.3.3 Interactions and nonlinear transforms

### Why It Matters
Products, bins, logs, and ratios can reveal patterns that raw variables hide. They should be added with a hypothesis, not by reflex.


```python
import pandas as pd

features = pd.DataFrame({"x1": [2, 4, 6], "x2": [10, 8, 7]})
features["x1_x2_interaction"] = features["x1"] * features["x2"]
features
```

## 6.3.4 Missing-value treatment

### Why It Matters
Imputation should preserve training-only information boundaries. Some missingness deserves its own indicator.


```python
import pandas as pd
from sklearn.impute import SimpleImputer

raw = pd.DataFrame({"spend": [10.0, None, 18.0, None]})
imputed = SimpleImputer(strategy="median").fit_transform(raw)
pd.DataFrame(imputed, columns=["spend_imputed"])
```
