## 6.1.1 Supervised versus unsupervised learning

### Why It Matters
Supervised learning predicts a labeled target, while unsupervised learning searches for structure without labels. Confusing the two leads to mismatched expectations.


```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs, make_classification
from sklearn.linear_model import LogisticRegression

X_sup, y_sup = make_classification(n_samples=120, n_features=4, n_informative=3, n_redundant=0, random_state=7)
clf = LogisticRegression(max_iter=1000).fit(X_sup, y_sup)

X_unsup, _ = make_blobs(n_samples=120, centers=3, random_state=8)
clusters = KMeans(n_clusters=3, n_init=20, random_state=8).fit_predict(X_unsup)

{
    "supervised_has_labels": True,
    "supervised_predictions_example": clf.predict(X_sup[:5]).tolist(),
    "unsupervised_has_labels": False,
    "unsupervised_cluster_ids_example": clusters[:5].tolist(),
}
```

## 6.1.3 Baseline-first mindset

### Why It Matters
A mean predictor, majority-class guess, or simple logistic model creates a floor that harder models must beat.


```python
import numpy as np
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

X = np.arange(40).reshape(-1, 1)
y = np.array([0] * 28 + [1] * 12)
model = DummyClassifier(strategy="most_frequent").fit(X, y)
preds = model.predict(X)
{"baseline_accuracy": round(float(accuracy_score(y, preds)), 3), "majority_class": int(preds[0])}
```

## 6.1.5 Training as an iterative loop

### Why It Matters
Data cleaning, feature design, evaluation, and error analysis are a loop rather than a one-shot fit.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import train_test_split

X, y = make_classification(n_samples=260, n_features=6, n_informative=4, random_state=9)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=9)
dummy = DummyClassifier(strategy="most_frequent").fit(X_train, y_train)
logit = LogisticRegression(max_iter=1000).fit(X_train, y_train)

history = {
    "baseline_accuracy": round(float(accuracy_score(y_test, dummy.predict(X_test))), 3),
    "logistic_accuracy": round(float(accuracy_score(y_test, logit.predict(X_test))), 3),
    "logistic_auc": round(float(roc_auc_score(y_test, logit.predict_proba(X_test)[:, 1])), 3),
}
history
```
