## 1. Truthiness Filter

**Techniques:** bool and truthiness, list

Solution:


```python
def keep_truthy(values):
    return [value for value in values if value]


sample = [0, 1, "", "python", [], [1, 2], None, True]
print(keep_truthy(sample))
```

## 2. String Normalization

**Techniques:** str and Unicode text

Solution:


```python
def normalize_names(raw: str) -> list[str]:
    return [part.strip().title() for part in raw.split(",") if part.strip()]


print(normalize_names(" alice,BOB,  cara de souza "))
```

## 3. Dictionary Inversion

**Techniques:** dict, defaultdict

Solution:


```python
from collections import defaultdict


def invert_cohorts(student_to_cohort: dict[str, str]) -> dict[str, list[str]]:
    grouped = defaultdict(list)
    for student, cohort in student_to_cohort.items():
        grouped[cohort].append(student)
    return dict(grouped)


records = {"Alice": "A", "Bob": "B", "Cara": "A", "Dan": "B"}
print(invert_cohorts(records))
```

## 4. Conditional Classification

**Techniques:** if / elif / else, comparison operators

Solution:


```python
def classify_score(score: int) -> str:
    if not 0 <= score <= 100:
        raise ValueError("score must be between 0 and 100")
    if score >= 70:
        return "distinction"
    if score >= 50:
        return "pass"
    return "fail"


for score in [92, 67, 48]:
    print(score, classify_score(score))
```

## 5. Pattern Matching Command Parser

**Techniques:** match / case

Solution:


```python
def describe_command(command: tuple) -> str:
    match command:
        case ("move", x, y):
            return f"move to ({x}, {y})"
        case ("say", message):
            return f"say: {message}"
        case ("quit",):
            return "quit program"
        case _:
            return "unknown command"


print(describe_command(("move", 10, 20)))
print(describe_command(("say", "hello")))
print(describe_command(("quit",)))
```

## 6. Keyword-Only Reporting Function

**Techniques:** parameters, arguments, defaults, keyword-only arguments

Solution:


```python
def format_report(name: str, score: float, *, rounded: bool = False) -> str:
    formatted = f"{score:.0f}" if rounded else f"{score:.2f}"
    return f"{name}: {formatted}"


print(format_report("Alice", 91.345))
print(format_report("Alice", 91.345, rounded=True))
```

## 7. Closure-Based Multiplier Factory

**Techniques:** closures and nonlocal, first-class functions

Solution:


```python
def make_running_total():
    total = 0

    def add(value: int | float):
        nonlocal total
        total += value
        return total

    return add


tracker = make_running_total()
print(tracker(5))
print(tracker(7))
print(tracker(-2))
```

## 8. Custom Iterable

**Techniques:** iteration protocol

Solution:


```python
class Countdown:
    def __init__(self, start: int):
        self.current = start

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        value = self.current
        self.current -= 1
        return value


print(list(Countdown(5)))
```

## 10. Safe Average with Exceptions

**Techniques:** raising exceptions well, assertions and invariants

Solution:


```python
def safe_average(values):
    if not values:
        raise ValueError("values must not be empty")
    if any(not isinstance(value, (int, float)) for value in values):
        raise TypeError("all values must be numeric")
    return sum(values) / len(values)


print(safe_average([10, 20, 30]))
```

## 13. Counter-Based Frequency Report

**Techniques:** collections.Counter

Solution:


```python
import re
from collections import Counter


def top_words(text: str, n: int = 3):
    words = re.findall(r"[a-zA-Z]+", text.lower())
    return Counter(words).most_common(n)


text = "Python is great, and Python is readable. Readable code is great."
print(top_words(text))
```

## 14. Iterator Pipeline

**Techniques:** iterator algebra with itertools, lazy vs eager evaluation

Solution:


```python
from itertools import islice


def first_five_even_squares():
    evens = filter(lambda value: value % 2 == 0, range(20))
    squares = map(lambda value: value * value, evens)
    return list(islice(squares, 5))


print(first_five_even_squares())
```

## 15. Protocol-Oriented Function

**Techniques:** Protocol and structural typing

Solution:


```python
from typing import Protocol

class Renderable(Protocol):
    def render(self) -> str: ...


def render_all(items: list[Renderable]) -> list[str]:
    return [item.render() for item in items]


class HtmlTitle:
    def render(self) -> str:
        return "<h1>Report</h1>"


class PlainLine:
    def render(self) -> str:
        return "plain text line"


print(render_all([HtmlTitle(), PlainLine()]))
```

## 16. Enum-Based State Machine

**Techniques:** enum

Solution:


```python
from enum import Enum

class Status(Enum):
    TODO = "todo"
    DOING = "doing"
    DONE = "done"


def next_status(status: Status) -> Status:
    mapping = {Status.TODO: Status.DOING, Status.DOING: Status.DONE, Status.DONE: Status.DONE}
    return mapping[status]


for item in Status:
    print(item, next_status(item))
```

## 17. Logging Wrapper Decorator

**Techniques:** decorators, logging

Solution:


```python
import logging

logging.basicConfig(level=logging.INFO, format="%(levelname)s:%(message)s")


def log_calls(func):
    def wrapper(*args, **kwargs):
        logging.info("calling %s", func.__name__)
        return func(*args, **kwargs)
    return wrapper


@log_calls
def add(a, b):
    return a + b


print(add(2, 3))
```

## 18. Dataclass Validation with Property

**Techniques:** properties, dataclasses

Solution:


```python
class Temperature:
    def __init__(self, celsius: float):
        self.celsius = celsius

    @property
    def celsius(self) -> float:
        return self._celsius

    @celsius.setter
    def celsius(self, value: float) -> None:
        if value < -273.15:
            raise ValueError("temperature below absolute zero")
        self._celsius = value


temp = Temperature(20)
print(temp.celsius)
temp.celsius = 25
print(temp.celsius)
try:
    temp.celsius = -300
except ValueError as exc:
    print(type(exc).__name__, exc)
```

## 19. Doctest-Friendly Utility

**Techniques:** doctests

Solution:


```python
def slugify(text: str) -> str:
    """Return a lowercase hyphenated slug.

    >>> slugify("Python Course")
    'python-course'
    >>> slugify("Data Science Notes")
    'data-science-notes'
    """
    return "-".join(text.lower().split())


print(slugify("Python Course Notes"))
```

## 20. Small Unit Test Suite

**Techniques:** unittest

Solution:


```python
import unittest


def is_even(n: int) -> bool:
    return n % 2 == 0


class TestIsEven(unittest.TestCase):
    def test_even_number(self):
        self.assertTrue(is_even(4))

    def test_odd_number(self):
        self.assertFalse(is_even(5))

    def test_negative_even_number(self):
        self.assertTrue(is_even(-2))


unittest.TextTestRunner().run(unittest.defaultTestLoader.loadTestsFromTestCase(TestIsEven))
```
