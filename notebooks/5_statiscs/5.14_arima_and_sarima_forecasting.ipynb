## 5.14.1 Stationarity intuition

### Why It Matters
ARIMA works best when the series has relatively stable statistical behavior after appropriate differencing.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(161)
series = pd.Series(np.linspace(40, 55, 18) + rng.normal(0, 1, 18))
differenced = series.diff().dropna()
pd.DataFrame({"original_head": series.head(6), "differenced_head": differenced.head(6)}).round(2)
```

## 5.14.2 AR and MA components

### Why It Matters
Autoregressive terms use previous values. Moving-average terms use previous shocks.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(270)
n = 120
ar = np.zeros(n)
ma = np.zeros(n)
ar_errors = rng.normal(0, 1, n)
ma_errors = rng.normal(0, 1, n)

for t in range(1, n):
    ar[t] = 0.75 * ar[t-1] + ar_errors[t]
    ma[t] = ma_errors[t] + 0.75 * ma_errors[t-1]

pd.DataFrame({
    "series": ["AR(1)", "MA(1)"],
    "lag_1_autocorr": [pd.Series(ar).autocorr(1), pd.Series(ma).autocorr(1)],
    "lag_2_autocorr": [pd.Series(ar).autocorr(2), pd.Series(ma).autocorr(2)],
}).round(3)
```

## 5.14.3 Differencing

### Why It Matters
Differencing removes trend-like persistence and helps stabilize the series.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(162)
series = pd.Series(np.linspace(10, 20, 12) + rng.normal(0, 0.8, 12))
pd.DataFrame({"value": series, "first_difference": series.diff()}).round(2)
```

## 5.14.4 Seasonal structure

### Why It Matters
SARIMA extends ARIMA by adding seasonal lags to repeated cycles.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(163)
months = pd.date_range("2021-01-01", periods=24, freq="M")
season = 4 * np.sin(np.arange(24) * 2 * np.pi / 12)
series = pd.Series(100 + season + rng.normal(0, 1, 24), index=months)
pd.DataFrame({"value": series, "seasonal_lag_12": series.shift(12)}).tail(6).round(2)
```

## 5.14.5 Forecast evaluation

### Why It Matters
Forecasts should be judged on future periods using error metrics such as MAE or RMSE, plus visual diagnostics.


```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error

actual = np.array([110, 113, 117, 119])
forecast = np.array([108, 114, 116, 121])
pd.DataFrame({
    "mae": [mean_absolute_error(actual, forecast)],
    "rmse": [mean_squared_error(actual, forecast) ** 0.5],
}).round(2)
```
