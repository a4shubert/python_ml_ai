## 5.11.1 Why regularize

### Why It Matters
Shrinkage reduces variance and helps prevent the model from overreacting to weak patterns.


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

rng = np.random.default_rng(260)
X = rng.normal(size=(220, 20))
y = 5 + 2 * X[:, 0] - 1.5 * X[:, 1] + rng.normal(0, 3, 220)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)

ols = LinearRegression().fit(X_train, y_train)
ridge = Ridge(alpha=2.0).fit(X_train, y_train)

pd.DataFrame({
    "model": ["OLS", "Ridge"],
    "test_rmse": [
        mean_squared_error(y_test, ols.predict(X_test)) ** 0.5,
        mean_squared_error(y_test, ridge.predict(X_test)) ** 0.5,
    ],
}).round(3)
```

## 5.11.2 Ridge and lasso

### Why It Matters
Ridge keeps all predictors but shrinks them. Lasso can set some coefficients to zero and simplify the feature set.


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso, Ridge

rng = np.random.default_rng(131)
x1 = rng.normal(size=180)
x2 = 0.88 * x1 + rng.normal(scale=0.25, size=180)
x3 = rng.normal(size=180)
y = 5 + 2.1 * x1 - 1.4 * x2 + 0.7 * x3 + rng.normal(0, 1.1, 180)
X = pd.DataFrame({"x1": x1, "x2": x2, "x3": x3})
pd.DataFrame({
    "ridge": Ridge(alpha=1.0).fit(X, y).coef_,
    "lasso": Lasso(alpha=0.08).fit(X, y).coef_,
}, index=X.columns).round(3)
```

## 5.11.3 Elastic net

### Why It Matters
Elastic net balances ridge-style stability with lasso-style sparsity when predictors are correlated.


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import ElasticNet

rng = np.random.default_rng(132)
X = pd.DataFrame(rng.normal(size=(160, 4)), columns=list("abcd"))
y = 4 + 2 * X["a"] - 1.5 * X["b"] + rng.normal(0, 1, 160)
model = ElasticNet(alpha=0.05, l1_ratio=0.5).fit(X, y)
pd.Series(model.coef_, index=X.columns).round(3)
```

## 5.11.4 Robust regression

### Why It Matters
Robust regression downweights unusual observations instead of letting them dominate the fit.


```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

rng = np.random.default_rng(133)
x = rng.normal(size=150)
y = 10 + 2.5 * x + rng.normal(0, 1, 150)
y[[10, 25, 90]] += [12, -15, 10]
X = sm.add_constant(x)
robust = sm.RLM(y, X).fit()
pd.Series(robust.params, index=["intercept", "x"]).round(3)
```

## 5.11.5 Prediction versus interpretation

### Why It Matters
Regularized models often improve predictive stability while changing how directly coefficients can be interpreted.


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

rng = np.random.default_rng(261)
X = rng.normal(size=(240, 12))
y = 10 + 3 * X[:, 0] - 2 * X[:, 1] + rng.normal(0, 4, 240)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=11)

ols = LinearRegression().fit(X_train, y_train)
ridge = Ridge(alpha=3.0).fit(X_train, y_train)

pd.DataFrame({
    "model": ["OLS", "Ridge"],
    "coef_l2_norm": [np.linalg.norm(ols.coef_), np.linalg.norm(ridge.coef_)],
    "test_mae": [
        mean_absolute_error(y_test, ols.predict(X_test)),
        mean_absolute_error(y_test, ridge.predict(X_test)),
    ],
}).round(3)
```
