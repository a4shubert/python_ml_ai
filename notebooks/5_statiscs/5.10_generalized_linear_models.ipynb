## 5.10.1 Why GLMs exist

### Why It Matters
A binary outcome should not be modeled as if it can take any real value. Count outcomes should not be modeled as if negative numbers are plausible.


```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

rng = np.random.default_rng(250)
x = rng.normal(0, 1, 220)
prob = 1 / (1 + np.exp(-(0.8 + 1.4 * x)))
y = rng.binomial(1, prob)

X = sm.add_constant(x)
ols = sm.OLS(y, X).fit()
ols_preds = ols.predict(X[:5])

logit = sm.Logit(y, X).fit(disp=False)
logit_preds = logit.predict(X[:5])

pd.DataFrame({
    "ols_prediction": ols_preds,
    "logit_prediction": logit_preds,
}).round(3)
```

## 5.10.2 Logistic regression

### Why It Matters
Logistic regression maps a linear score onto probabilities through the logit link.


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

rng = np.random.default_rng(121)
df = pd.DataFrame({
    "response_time": rng.normal(12, 3, 240),
    "priority": rng.binomial(1, 0.35, 240),
})
logits = -0.6 - 0.18 * df["response_time"] + 1.0 * df["priority"]
df["renewed"] = rng.binomial(1, 1 / (1 + np.exp(-logits)))
model = smf.logit("renewed ~ response_time + priority", data=df).fit(disp=False)
model.predict(df.head()).round(3).tolist()
```

## 5.10.3 Poisson modeling intuition

### Why It Matters
Poisson models start with counts whose expected rate changes with predictors.


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

rng = np.random.default_rng(122)
df = pd.DataFrame({
    "visits": rng.normal(5, 1.5, 220),
    "priority": rng.binomial(1, 0.3, 220),
})
rate = np.exp(0.8 + 0.15 * df["visits"] + 0.4 * df["priority"])
df["tickets"] = rng.poisson(rate)
model = smf.poisson("tickets ~ visits + priority", data=df).fit(disp=False)
model.predict(df.head()).round(2).tolist()
```

## 5.10.5 Interpretation on the right scale

### Why It Matters
Stakeholders often understand probabilities and expected counts better than logits or log-rates.


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

rng = np.random.default_rng(251)
df = pd.DataFrame({
    "response_time": rng.normal(12, 3, 200),
    "priority": rng.binomial(1, 0.35, 200),
})
logits = -1.0 - 0.2 * df["response_time"] + 1.2 * df["priority"]
df["renewed"] = rng.binomial(1, 1 / (1 + np.exp(-logits)))
model = smf.logit("renewed ~ response_time + priority", data=df).fit(disp=False)

example = df.iloc[[0, 1, 2]].copy()
example["predicted_probability"] = model.predict(example)
example[["response_time", "priority", "predicted_probability"]].round(3)
```
