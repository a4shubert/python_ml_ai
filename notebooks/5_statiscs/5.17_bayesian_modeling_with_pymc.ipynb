## 5.17.2 Posterior sampling

### Why It Matters
Sampling approximates the posterior with many draws, making uncertainty visible instead of hidden.


```python
import numpy as np
import pymc as pm

observed = np.r_[np.ones(20), np.zeros(8)]
with pm.Model() as model:
    theta = pm.Beta("theta", 2, 2)
    pm.Bernoulli("obs", p=theta, observed=observed)
    trace = pm.sample(draws=250, tune=250, chains=2, cores=1, progressbar=False, random_seed=42)
trace.posterior["theta"].mean().values.round(3)
```

## 5.17.3 Posterior summaries

### Why It Matters
Means, medians, intervals, and trace plots reveal different aspects of posterior behavior.


```python
import arviz as az
import numpy as np
import pymc as pm

observed = np.r_[np.ones(18), np.zeros(7)]
with pm.Model() as model:
    theta = pm.Beta("theta", 2, 2)
    pm.Bernoulli("obs", p=theta, observed=observed)
    trace = pm.sample(draws=250, tune=250, chains=2, cores=1, progressbar=False, random_seed=42)
az.summary(trace, var_names=["theta"])
```

## 5.17.4 Posterior predictive checks

### Why It Matters
A good Bayesian model should generate data that looks like the data you actually observed.


```python
import arviz as az
import numpy as np
import pymc as pm

observed = np.r_[np.ones(22), np.zeros(10)]
with pm.Model() as model:
    theta = pm.Beta("theta", 2, 2)
    pm.Bernoulli("obs", p=theta, observed=observed)
    trace = pm.sample(draws=250, tune=250, chains=2, cores=1, progressbar=False, random_seed=42)
    ppc = pm.sample_posterior_predictive(trace, progressbar=False, random_seed=42)
az.summary(ppc, var_names=["obs"])
```

## 5.17.5 Diagnostics

### Why It Matters
R-hat, effective sample size, and divergences tell you whether the computation is trustworthy.


```python
import arviz as az
import numpy as np
import pymc as pm

observed = np.r_[np.ones(19), np.zeros(9)]
with pm.Model() as model:
    theta = pm.Beta("theta", 2, 2)
    pm.Bernoulli("obs", p=theta, observed=observed)
    trace = pm.sample(draws=250, tune=250, chains=2, cores=1, progressbar=False, random_seed=42)
az.rhat(trace).to_dataframe().round(3)
```

## 5.17.6 Start small

### Why It Matters
