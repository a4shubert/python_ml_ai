## 5.9.1 Linearity

### Why It Matters
Residual plots help you see whether a straight-line relationship is a reasonable summary or whether curvature remains.


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

rng = np.random.default_rng(111)
x = rng.uniform(0, 10, 220)
y = 3 + 1.5 * x + rng.normal(0, 1 + 0.1 * x, 220)
df = pd.DataFrame({"x": x, "y": y})
model = smf.ols("y ~ x", data=df).fit()
pd.DataFrame({"fitted": model.fittedvalues.head(), "residual": model.resid.head()}).round(2)
```

## 5.9.2 Homoscedasticity

### Why It Matters
If residual spread changes dramatically with the fitted values, standard errors and intervals become less trustworthy.


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

rng = np.random.default_rng(112)
x = rng.uniform(0, 8, 240)
y = 5 + 2 * x + rng.normal(0, 0.8 + 0.4 * x, 240)
df = pd.DataFrame({"x": x, "y": y})
model = smf.ols("y ~ x", data=df).fit()
pd.DataFrame({"fitted": model.fittedvalues, "abs_residual": abs(model.resid)}).corr().round(2)
```

## 5.9.3 Normality of residuals

### Why It Matters
Residual normality matters more for small-sample inference than for many predictive tasks, but extreme departures still deserve attention.


```python
import numpy as np
from scipy.stats import skew, kurtosis

rng = np.random.default_rng(113)
residual_like = rng.normal(0, 1, 300)
{"skew": round(float(skew(residual_like)), 3), "kurtosis": round(float(kurtosis(residual_like)), 3)}
```

## 5.9.4 Multicollinearity

### Why It Matters
Correlated predictors can make coefficient estimates unstable and inflate standard errors.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(114)
x1 = rng.normal(size=180)
x2 = 0.92 * x1 + rng.normal(scale=0.2, size=180)
x3 = rng.normal(size=180)
pd.DataFrame({"x1": x1, "x2": x2, "x3": x3}).corr().round(2)
```

## 5.9.5 Influence and leverage

### Why It Matters
Some points matter because they are unusual in predictor space, not just because the outcome is extreme.


```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

rng = np.random.default_rng(115)
x = rng.normal(size=160)
y = 20 + 3 * x + rng.normal(0, 1, 160)
y[5] += 18
y[90] -= 20
X = sm.add_constant(x)
model = sm.OLS(y, X).fit()
influence = model.get_influence().cooks_distance[0]
pd.Series(influence).sort_values(ascending=False).head().round(3)
```

## 5.9.6 Revision is part of the workflow

### Why It Matters
Diagnostics should lead to action: transform, simplify, add structure, or narrow the question.


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

rng = np.random.default_rng(240)
x = rng.uniform(-3, 3, 250)
y = 8 + 2.5 * x + 1.2 * x**2 + rng.normal(0, 1.2, 250)
df = pd.DataFrame({"x": x, "y": y})

linear = smf.ols("y ~ x", data=df).fit()
quadratic = smf.ols("y ~ x + I(x**2)", data=df).fit()

pd.DataFrame({
    "model": ["linear", "quadratic"],
    "aic": [linear.aic, quadratic.aic],
    "r_squared": [linear.rsquared, quadratic.rsquared],
}).round(3)
```
