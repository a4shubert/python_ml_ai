## 5.3.1 Distribution shape

### Why It Matters
Histograms, quantiles, and boxplots tell you whether a variable is symmetric, skewed, heavy-tailed, or multimodal. Those shape differences influence summaries and model families.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(51)
spend = pd.Series(rng.lognormal(mean=3.2, sigma=0.5, size=250), name="marketing_spend")
spend.describe().round(2)
```

## 5.3.2 Outliers and leverage points

### Why It Matters
Some unusual points are data problems. Others are the reason the dataset is interesting. EDA helps you distinguish those cases before deleting anything.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(52)
values = pd.Series(rng.normal(40, 6, 200), name="delivery_time")
values.iloc[[3, 44, 150]] = [12, 88, 97]
values.sort_values().tail(10).reset_index(drop=True)
```

## 5.3.3 Missingness patterns

### Why It Matters
A missing value is often information about a process breakdown, a skipped question, or a subgroup. Missingness deserves its own analysis.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(53)
df = pd.DataFrame({
    "region": rng.choice(["north", "south", "east", "west"], size=120),
    "score": rng.normal(7.5, 1.0, size=120),
})
df.loc[df["region"].isin(["south", "west"]), "score"] = np.nan
df.isna().groupby(df["region"]).mean().round(2)
```

## 5.3.4 Pairwise structure

### Why It Matters
Scatterplots and grouped summaries show whether a relationship is linear, curved, segmented, or dominated by a few cases.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(54)
x = rng.normal(50, 10, 180)
y = 10 + 0.6 * x + rng.normal(0, 4, 180)
pd.DataFrame({"ad_spend": x, "sales": y}).corr().round(2)
```

## 5.3.5 Transformations

### Why It Matters
Logs, roots, and winsorization are not cosmetic. They are tools for stabilizing scale, revealing patterns, and making comparisons fairer.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(55)
raw = pd.Series(rng.lognormal(mean=2.8, sigma=0.9, size=180), name="claim_amount")
transformed = np.log(raw)
pd.DataFrame({
    "raw_mean": [raw.mean()],
    "raw_std": [raw.std()],
    "log_mean": [transformed.mean()],
    "log_std": [transformed.std()],
}).round(2)
```
