## 5.16.1 Prior, likelihood, posterior

### Why It Matters
The prior captures what was plausible before the current data, the likelihood measures compatibility with the observed data, and the posterior combines the two.


```python
import numpy as np
import pandas as pd
from scipy.stats import binom

theta_grid = np.linspace(0.01, 0.99, 99)
prior = np.ones_like(theta_grid)
likelihood = binom.pmf(8, 12, theta_grid)
posterior = prior * likelihood
posterior = posterior / posterior.sum()

pd.DataFrame({
    "theta": theta_grid,
    "prior": prior / prior.sum(),
    "likelihood_scaled": likelihood / likelihood.sum(),
    "posterior": posterior,
}).head(10).round(4)
```

## 5.16.2 Credible intervals

### Why It Matters
A Bayesian credible interval is directly about plausible parameter values under the posterior. That is different from the long-run repeated-sampling interpretation of a confidence interval.


```python
import numpy as np

rng = np.random.default_rng(181)
posterior = rng.beta(40, 18, size=6000)
tuple(round(x, 3) for x in np.quantile(posterior, [0.025, 0.5, 0.975]))
```

## 5.16.3 Weakly informative priors

### Why It Matters
Weakly informative priors are often useful because they stabilize models without pretending you know too much.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(182)
weak_prior = rng.beta(2, 2, size=5000)
stronger_prior = rng.beta(8, 4, size=5000)
pd.DataFrame({
    "weak_prior_mean": [weak_prior.mean()],
    "stronger_prior_mean": [stronger_prior.mean()],
}).round(3)
```

## 5.16.4 Posterior updating

### Why It Matters
Bayesian reasoning mirrors how many people already think informally: beliefs should move when evidence accumulates.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(183)
prior = rng.beta(2, 2, size=4000)
posterior = rng.beta(2 + 26, 2 + 10, size=4000)
pd.DataFrame({
    "distribution": ["prior", "posterior"],
    "mean": [prior.mean(), posterior.mean()],
}).round(3)
```

## 5.16.5 Posterior predictive thinking

### Why It Matters
The posterior becomes most useful when you convert it into predictions or decision-relevant quantities.


```python
import numpy as np

rng = np.random.default_rng(184)
posterior = rng.beta(34, 16, size=5000)
predicted_conversion = rng.binomial(1, posterior).mean()
round(float(predicted_conversion), 3)
```

## 5.16.6 Frequentist versus Bayesian framing

### Why It Matters
The same data can often be analyzed either way, but the language of uncertainty and prior information changes.


```python
import numpy as np
import pandas as pd
from scipy import stats

rng = np.random.default_rng(290)
data = rng.binomial(1, 0.68, size=60)
p_hat = data.mean()
se = np.sqrt(p_hat * (1 - p_hat) / len(data))
ci = stats.norm.interval(0.95, loc=p_hat, scale=se)

posterior = rng.beta(2 + data.sum(), 2 + len(data) - data.sum(), size=6000)
bayes_interval = np.quantile(posterior, [0.025, 0.975])

pd.DataFrame({
    "method": ["frequentist_CI", "bayesian_credible_interval"],
    "lower": [ci[0], bayes_interval[0]],
    "upper": [ci[1], bayes_interval[1]],
}).round(3)
```
