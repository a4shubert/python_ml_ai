## 5.8.1 Simple regression intuition

### Why It Matters
With one predictor, the slope describes the average directional change in the expected outcome.


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

rng = np.random.default_rng(101)
x = rng.normal(0, 1, 200)
y = 50 + 4.5 * x + rng.normal(0, 3, 200)
df = pd.DataFrame({"x": x, "y": y})
smf.ols("y ~ x", data=df).fit().params.round(2)
```

## 5.8.2 Multiple regression

### Why It Matters
Adding predictors lets you estimate partial relationships while adjusting for the other included variables.


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

rng = np.random.default_rng(102)
df = pd.DataFrame({
    "study": rng.normal(7, 2, 220),
    "attendance": rng.normal(0.85, 0.07, 220),
    "prior": rng.normal(72, 8, 220),
})
df["score"] = 20 + 2.5 * df["study"] + 16 * df["attendance"] + 0.4 * df["prior"] + rng.normal(0, 4, 220)
smf.ols("score ~ study + attendance + prior", data=df).fit().params.round(2)
```

## 5.8.3 Intercepts and centering

### Why It Matters
The intercept is only meaningful if zero is meaningful. Centering can improve interpretation without changing the fitted relationships.


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

rng = np.random.default_rng(103)
hours = rng.normal(8, 2, 180)
centered_hours = hours - hours.mean()
score = 55 + 3.2 * hours + rng.normal(0, 5, 180)
df = pd.DataFrame({"hours": hours, "hours_centered": centered_hours, "score": score})
model = smf.ols("score ~ hours_centered", data=df).fit()
model.params.round(2)
```

## 5.8.4 Residuals

### Why It Matters
Residuals are the model errors that remain after the fitted linear relationship has done its work.


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

rng = np.random.default_rng(104)
x = rng.normal(0, 1, 150)
y = 10 + 2.2 * x + rng.normal(0, 2, 150)
df = pd.DataFrame({"x": x, "y": y})
model = smf.ols("y ~ x", data=df).fit()
pd.DataFrame({"fitted": model.fittedvalues.head(), "residual": model.resid.head()}).round(2)
```

## 5.8.5 Goodness of fit

### Why It Matters
R-squared is useful but incomplete. A model can explain variance and still be badly specified.


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

rng = np.random.default_rng(105)
x = rng.normal(0, 1, 160)
y = 15 + 5 * x + rng.normal(0, 4, 160)
df = pd.DataFrame({"x": x, "y": y})
model = smf.ols("y ~ x", data=df).fit()
{"r_squared": round(float(model.rsquared), 3), "adj_r_squared": round(float(model.rsquared_adj), 3)}
```

## 5.8.6 Association is not causation

### Why It Matters
Regression coefficients are conditional associations unless the design and assumptions support stronger claims.


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

rng = np.random.default_rng(230)
aptitude = rng.normal(0, 1, 300)
study_hours = 6 + 1.8 * aptitude + rng.normal(0, 0.8, 300)
exam_score = 68 + 5.5 * aptitude + 2.0 * study_hours + rng.normal(0, 3, 300)
df = pd.DataFrame({"aptitude": aptitude, "study_hours": study_hours, "exam_score": exam_score})

naive = smf.ols("exam_score ~ study_hours", data=df).fit()
adjusted = smf.ols("exam_score ~ study_hours + aptitude", data=df).fit()

pd.DataFrame({
    "model": ["naive", "adjusted_for_aptitude"],
    "study_hours_coef": [naive.params["study_hours"], adjusted.params["study_hours"]],
}).round(3)
```
