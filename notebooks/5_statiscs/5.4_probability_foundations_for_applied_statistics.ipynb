## 5.4.1 Events and sample spaces

### Why It Matters
A sample space lists possible outcomes, while an event selects outcomes relevant to a question. This matters because business questions are usually event questions.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(210)
outcomes = pd.Series(rng.choice(["low", "medium", "high"], p=[0.25, 0.5, 0.25], size=5000), name="outcome")
event_prob = (outcomes.isin(["medium", "high"])).mean()
outcomes.value_counts(normalize=True).rename("probability").round(3).to_frame().assign(event_medium_or_high=event_prob)
```

## 5.4.4 Random variables and expectation

### Why It Matters
A random variable translates outcomes into numbers. Its expectation describes a long-run average, while variance describes the instability around that average.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(61)
draws = rng.normal(loc=100, scale=15, size=1000)
pd.DataFrame({
    "expectation_estimate": [draws.mean()],
    "variance_estimate": [draws.var()],
}).round(2)
```

## 5.4.5 Discrete and continuous models

### Why It Matters
Counts, proportions, and physical measurements behave differently. The right probability family keeps your assumptions aligned with the support of the data.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(211)
discrete_counts = pd.Series(rng.poisson(lam=3.5, size=500), name="tickets")
continuous_wait = pd.Series(rng.normal(loc=12, scale=2.5, size=500), name="wait_minutes")

pd.DataFrame({
    "discrete_unique_values": [int(discrete_counts.nunique())],
    "discrete_min": [int(discrete_counts.min())],
    "discrete_max": [int(discrete_counts.max())],
    "continuous_mean": [continuous_wait.mean()],
    "continuous_std": [continuous_wait.std()],
}).round(2)
```
