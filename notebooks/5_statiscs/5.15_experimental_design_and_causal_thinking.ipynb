## 5.15.2 Treatment and control

### Why It Matters
The control group defines the counterfactual you would like to compare against the treatment group.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(280)
n = 400
treatment = rng.binomial(1, 0.5, n)
outcome = 72 + 3 * treatment + rng.normal(0, 6, n)
experiment = pd.DataFrame({"treatment": treatment, "outcome": outcome})
experiment.groupby("treatment")["outcome"].agg(["mean", "count"]).round(2)
```

## 5.15.5 Internal versus external validity

### Why It Matters
A result can be credible inside the experiment and still fail to generalize to other settings or populations.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(281)
urban_treat = rng.normal(78, 8, 180)
urban_ctrl = rng.normal(74, 8, 180)
rural_treat = rng.normal(69, 8, 180)
rural_ctrl = rng.normal(68, 8, 180)

pd.DataFrame({
    "population": ["urban_experiment", "rural_new_market"],
    "estimated_effect": [urban_treat.mean() - urban_ctrl.mean(), rural_treat.mean() - rural_ctrl.mean()],
}).round(2)
```

## 5.15.6 Causal humility

### Why It Matters
Strong language should be reserved for designs that genuinely support it.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(282)
experience = rng.normal(0, 1, 500)
treated_obs = rng.binomial(1, 1 / (1 + np.exp(-(1.1 * experience))))
outcome_obs = 70 + 5 * experience + 2.5 * treated_obs + rng.normal(0, 2, 500)
naive_obs_effect = outcome_obs[treated_obs == 1].mean() - outcome_obs[treated_obs == 0].mean()

treated_rct = rng.binomial(1, 0.5, 500)
outcome_rct = 70 + 5 * experience + 2.5 * treated_rct + rng.normal(0, 2, 500)
rct_effect = outcome_rct[treated_rct == 1].mean() - outcome_rct[treated_rct == 0].mean()

pd.DataFrame({
    "estimate_type": ["naive_observational", "randomized_experiment"],
    "estimated_effect": [naive_obs_effect, rct_effect],
}).round(2)
```
