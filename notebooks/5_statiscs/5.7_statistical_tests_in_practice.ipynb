## 5.7.1 Difference-in-means tests

### Why It Matters
Use t-style tests for numeric outcomes when the design is about mean differences and independence is credible.


```python
import numpy as np
from scipy.stats import ttest_ind

rng = np.random.default_rng(91)
control = rng.normal(68, 7, 80)
treatment = rng.normal(71, 7, 85)
_, p_value = ttest_ind(control, treatment, equal_var=False)
round(float(p_value), 4)
```

## 5.7.2 Proportion tests

### Why It Matters
When the target is conversion, default rate, or success probability, a proportion test is often more natural than a mean-based test.


```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

successes = np.array([58, 73])
trials = np.array([150, 150])
stat, p_value = proportions_ztest(successes, trials)
{"z_stat": round(float(stat), 3), "p_value": round(float(p_value), 4)}
```

## 5.7.3 Chi-square tests

### Why It Matters
Chi-square procedures test whether categorical variables appear statistically associated.


```python
import pandas as pd
from scipy.stats import chi2_contingency

table = pd.DataFrame([[35, 18], [22, 29]], index=["A", "B"], columns=["renewed", "not_renewed"])
chi2, p_value, _, _ = chi2_contingency(table)
{"chi2": round(float(chi2), 3), "p_value": round(float(p_value), 4)}
```

## 5.7.4 Nonparametric alternatives

### Why It Matters
Rank-based tests are attractive when outliers or strong non-normality make parametric assumptions questionable.


```python
import numpy as np
from scipy.stats import mannwhitneyu

rng = np.random.default_rng(92)
a = rng.lognormal(mean=3.0, sigma=0.5, size=60)
b = rng.lognormal(mean=3.2, sigma=0.6, size=65)
stat, p_value = mannwhitneyu(a, b, alternative="two-sided")
{"u_stat": round(float(stat), 2), "p_value": round(float(p_value), 4)}
```

## 5.7.5 Assumption checks

### Why It Matters
Expected counts, approximate independence, and outcome scale should be checked before treating a p-value as trustworthy.


```python
import pandas as pd
from scipy.stats import chi2_contingency

table = pd.DataFrame(
    [[42, 10, 6], [18, 22, 15]],
    index=["A", "B"],
    columns=["low", "mid", "high"],
)
chi2, p_value, _, expected = chi2_contingency(table)
expected_df = pd.DataFrame(expected, index=table.index, columns=table.columns).round(2)
{
    "minimum_expected_count": float(expected_df.min().min()),
    "chi_square_p_value": round(float(p_value), 4),
}
```

## 5.7.6 Test selection is design selection

### Why It Matters
Once you know outcome type, number of groups, and whether samples are paired, the right test is usually easier to identify.


```python
import pandas as pd

def choose_test(outcome_type, groups, paired=False):
    if outcome_type == "numeric" and groups == 2 and not paired:
        return "Welch t-test"
    if outcome_type == "binary" and groups == 2 and not paired:
        return "Two-proportion z-test"
    if outcome_type == "categorical" and groups == 2:
        return "Chi-square test"
    return "Needs a more specific design decision"

scenarios = pd.DataFrame({
    "question": [
        "compare average spend between two campaigns",
        "compare conversion rates for two landing pages",
        "test association between region and satisfaction band",
    ],
    "recommended_test": [
        choose_test("numeric", 2),
        choose_test("binary", 2),
        choose_test("categorical", 2),
    ],
})
scenarios
```
