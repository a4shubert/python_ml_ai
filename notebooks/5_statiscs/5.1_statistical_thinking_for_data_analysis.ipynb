## 5.1.1 Population, sample, and target of inference

### Why It Matters
A sample is what you observed, but the target of inference is usually broader: future customers, similar patients, next quarter, or a manufacturing process. A careful analyst states that target explicitly before computing summaries.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(42)
population = pd.DataFrame({
    "weekly_revenue": rng.normal(75, 12, size=5000),
})
sample = population.sample(120, random_state=7)

pd.DataFrame({
    "population_mean": [population["weekly_revenue"].mean()],
    "sample_mean": [sample["weekly_revenue"].mean()],
    "sample_size": [len(sample)],
}).round(2)
```

## 5.1.2 Variability is part of the signal

### Why It Matters
A mean without spread is incomplete. Two groups can share the same average while having very different risk, volatility, or operational predictability.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(43)
stable = rng.normal(72, 6, size=150)
volatile = rng.normal(72, 18, size=150)
df = pd.DataFrame({
    "segment": ["stable"] * 150 + ["volatile"] * 150,
    "value": np.concatenate([stable, volatile]),
})

df.groupby("segment")["value"].agg(["mean", "std", "min", "max"]).round(2)
```

## 5.1.3 Estimation before testing

### Why It Matters
A disciplined workflow asks what the effect size looks like before asking whether it is statistically detectable. That keeps attention on magnitude and practical relevance.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(44)
control = rng.normal(71, 8, size=100)
treatment = rng.normal(74, 8, size=100)
point_estimate = treatment.mean() - control.mean()

pd.DataFrame({
    "control_mean": [control.mean()],
    "treatment_mean": [treatment.mean()],
    "difference_in_means": [point_estimate],
}).round(2)
```

## 5.1.4 Inference versus prediction

### Why It Matters
Inference emphasizes explanation and uncertainty around parameters. Prediction emphasizes accuracy on new data. Many teams confuse the two and end up using the wrong success metric.


```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

rng = np.random.default_rng(45)
X = rng.normal(size=(240, 1))
y = 50 + 6 * X[:, 0] + rng.normal(scale=3, size=240)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=7)
model = LinearRegression().fit(X_train, y_train)
preds = model.predict(X_test)

{
    "slope_for_inference": round(model.coef_[0], 2),
    "test_mae_for_prediction": round(mean_absolute_error(y_test, preds), 2),
}
```

## 5.1.5 Baselines anchor judgment

### Why It Matters
Every estimate should be compared with a naive baseline such as a historical average, a no-change assumption, or a majority-class guess. Baselines keep small improvements from being oversold.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(46)
revenue = pd.Series(rng.normal(82, 10, size=52))
baseline = revenue.mean()
next_month_projection = baseline

pd.DataFrame({
    "historical_average_baseline": [baseline],
    "example_forecast_for_next_month": [next_month_projection],
}).round(2)
```
