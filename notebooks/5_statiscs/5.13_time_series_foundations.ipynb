## 5.13.2 Trend and seasonality

### Why It Matters
Trend is long-run movement; seasonality is repeated cyclical structure such as weekly, monthly, or quarterly effects.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(151)
months = pd.date_range("2020-01-01", periods=24, freq="M")
trend = np.linspace(100, 120, 24)
season = 5 * np.sin(np.arange(24) * 2 * np.pi / 12)
series = pd.Series(trend + season + rng.normal(0, 1, 24), index=months)
pd.DataFrame({"trend_like_series": series}).head(12).round(2)
```

## 5.13.3 Lag dependence

### Why It Matters
Past values often help predict future values because systems carry momentum, inertia, or delayed feedback.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(152)
series = pd.Series(rng.normal(100, 3, 15))
pd.DataFrame({"value": series, "lag_1": series.shift(1), "lag_2": series.shift(2)}).head(8).round(2)
```

## 5.13.4 Rolling summaries

### Why It Matters
Rolling means and rolling variability expose local changes in level and volatility.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(153)
series = pd.Series(rng.normal(40, 4, 20))
pd.DataFrame({
    "value": series,
    "rolling_mean_3": series.rolling(3).mean(),
    "rolling_std_5": series.rolling(5).std(),
}).tail(8).round(2)
```

## 5.13.5 Autocorrelation

### Why It Matters
Autocorrelation quantifies how related the series is to its own recent past.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(154)
series = pd.Series(rng.normal(0, 1, 100)).cumsum()
pd.DataFrame({
    "lag_1_autocorr": [series.autocorr(lag=1)],
    "lag_6_autocorr": [series.autocorr(lag=6)],
}).round(3)
```
