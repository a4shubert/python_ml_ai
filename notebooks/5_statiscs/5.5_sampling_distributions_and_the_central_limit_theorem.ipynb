## 5.5.1 Why one estimate is not enough

### Why It Matters
A sample mean is not fixed truth. It is one draw from a distribution of possible sample means.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(71)
population = rng.normal(82, 14, size=10000)
estimates = [rng.choice(population, size=60, replace=False).mean() for _ in range(12)]
pd.Series(estimates, name="sample_mean").round(2)
```

## 5.5.2 Standard error

### Why It Matters
The standard error measures the expected fluctuation of an estimator across repeated samples. It shrinks with larger samples, but only if the data remains representative.


```python
import numpy as np

rng = np.random.default_rng(72)
sample = rng.normal(50, 9, size=80)
standard_error = sample.std(ddof=1) / np.sqrt(len(sample))
round(float(standard_error), 3)
```

## 5.5.3 Central Limit Theorem intuition

### Why It Matters
Averages tend to stabilize and become approximately bell-shaped even when the raw data is not. That makes interval estimation possible in many messy settings.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(73)
bootstrap_means = [rng.exponential(scale=20, size=120).mean() for _ in range(600)]
pd.Series(bootstrap_means).describe()[["mean", "std", "25%", "50%", "75%"]].round(2)
```

## 5.5.4 Bootstrap thinking

### Why It Matters
Bootstrapping approximates estimator variability by resampling the observed data. It is useful when analytic formulas are awkward or unavailable.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(74)
sample = rng.lognormal(mean=3.0, sigma=0.7, size=90)
boot = [rng.choice(sample, size=len(sample), replace=True).mean() for _ in range(400)]
pd.DataFrame({"bootstrap_mean": boot}).describe().round(2)
```

## 5.5.5 Sample size and stability

### Why It Matters
Larger samples reduce noise but do not cure bias. More data from a distorted process is still distorted.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(75)
pop = rng.normal(100, 20, size=50000)
se_small = np.std([rng.choice(pop, 30, replace=False).mean() for _ in range(300)], ddof=1)
se_large = np.std([rng.choice(pop, 180, replace=False).mean() for _ in range(300)], ddof=1)
pd.DataFrame({"sample_size": [30, 180], "empirical_standard_error": [se_small, se_large]}).round(2)
```
