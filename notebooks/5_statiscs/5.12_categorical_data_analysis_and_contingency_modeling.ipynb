## 5.12.1 Contingency tables

### Why It Matters
A contingency table is the starting point for many categorical questions because it summarizes the joint count pattern directly.


```python
import pandas as pd

table = pd.DataFrame(
    [[48, 15, 7], [31, 24, 19]],
    index=["web", "store"],
    columns=["low", "mid", "high"],
)
table
```

## 5.12.2 Independence and association

### Why It Matters
The core question is often whether two categorical variables look statistically independent or not.


```python
import pandas as pd
from scipy.stats import chi2_contingency

table = pd.DataFrame(
    [[48, 15, 7], [31, 24, 19]],
    index=["web", "store"],
    columns=["low", "mid", "high"],
)
_, p_value, _, _ = chi2_contingency(table)
round(float(p_value), 4)
```

## 5.12.3 Odds ratios

### Why It Matters
Odds ratios appear repeatedly in logistic models and case-control studies because they summarize relative event likelihood compactly.


```python
import pandas as pd

odds = pd.DataFrame({
    "group": ["control", "treatment"],
    "renewed": [42, 61],
    "not_renewed": [58, 39],
})
odds["odds"] = odds["renewed"] / odds["not_renewed"]
odds
```
