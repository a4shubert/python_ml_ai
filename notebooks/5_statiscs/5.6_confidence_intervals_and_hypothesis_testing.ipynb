## 5.6.2 Confidence intervals

### Why It Matters
A confidence interval comes from a procedure that captures the true parameter at a stated long-run rate. In practice it gives a scale for how uncertain you still are.


```python
import numpy as np
from scipy import stats

rng = np.random.default_rng(81)
x = rng.normal(74, 10, size=120)
mean = x.mean()
se = x.std(ddof=1) / np.sqrt(len(x))
interval = stats.norm.interval(0.95, loc=mean, scale=se)
tuple(round(v, 2) for v in interval)
```

## 5.6.3 p-values

### Why It Matters
A p-value measures how surprising the observed data would be if the null model were true. It is not the probability that the null is true.


```python
import numpy as np
from scipy.stats import ttest_ind

rng = np.random.default_rng(82)
a = rng.normal(70, 9, size=90)
b = rng.normal(73, 9, size=90)
_, p_value = ttest_ind(a, b, equal_var=False)
round(float(p_value), 5)
```

## 5.6.4 Type I and Type II errors

### Why It Matters
False alarms and missed detections are cost tradeoffs. There is no universally correct balance.


```python
import numpy as np
import pandas as pd
from scipy.stats import norm

rng = np.random.default_rng(220)
alpha = 0.05
runs = 2000
false_positives = 0
missed_detections = 0
true_effect = 1.5

for _ in range(runs):
    a = rng.normal(70, 8, 60)
    b_null = rng.normal(70, 8, 60)
    b_effect = rng.normal(70 + true_effect, 8, 60)

    diff_null = b_null.mean() - a.mean()
    se_null = np.sqrt(b_null.var(ddof=1)/len(b_null) + a.var(ddof=1)/len(a))
    p_null = 2 * (1 - norm.cdf(abs(diff_null / se_null)))
    false_positives += p_null < alpha

    diff_eff = b_effect.mean() - a.mean()
    se_eff = np.sqrt(b_effect.var(ddof=1)/len(b_effect) + a.var(ddof=1)/len(a))
    p_eff = 2 * (1 - norm.cdf(abs(diff_eff / se_eff)))
    missed_detections += p_eff >= alpha

pd.DataFrame({
    "type_i_error_rate": [false_positives / runs],
    "type_ii_error_rate": [missed_detections / runs],
}).round(3)
```

## 5.6.6 Multiplicity and discipline

### Why It Matters
When you ask many questions of one dataset, spurious significance becomes more likely. Label exploratory results honestly.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(83)
p_values = pd.Series(rng.uniform(0, 0.2, size=15), name="p_value")
discoveries = (p_values < 0.05).sum()
pd.DataFrame({"tests_run": [len(p_values)], "p_values_below_0.05": [int(discoveries)]})
```
