## 5.2.2 Observational versus experimental data

### Why It Matters
Observational data records the world as it happened. Experimental data changes one part of the world intentionally. The stronger the intervention, the stronger the causal story you can usually tell.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(200)
n = 600
experience = rng.normal(0, 1, n)
treatment_prob = 1 / (1 + np.exp(-(1.2 * experience)))
treatment = rng.binomial(1, treatment_prob)
outcome = 70 + 4 * experience + 3 * treatment + rng.normal(0, 2, n)

observational = pd.DataFrame({"experience": experience, "treatment": treatment, "outcome": outcome})
naive_diff = observational.groupby("treatment")["outcome"].mean().diff().iloc[-1]

randomized_treatment = rng.binomial(1, 0.5, n)
randomized_outcome = 70 + 4 * experience + 3 * randomized_treatment + rng.normal(0, 2, n)
randomized = pd.DataFrame({"treatment": randomized_treatment, "outcome": randomized_outcome})
randomized_diff = randomized.groupby("treatment")["outcome"].mean().diff().iloc[-1]

pd.DataFrame({
    "design": ["observational", "randomized"],
    "estimated_treatment_effect": [naive_diff, randomized_diff],
}).round(2)
```

## 5.2.5 Operational definitions

### Why It Matters
Whenever you define churn, quality, fraud, or engagement, you are making a modeling choice. That definition controls which examples the model sees.


```python
import pandas as pd

orders = pd.DataFrame({
    "customer_id": [101, 102, 103, 104, 105],
    "days_since_last_order": [12, 97, 35, 141, 63],
    "orders_last_90_days": [4, 0, 2, 0, 1],
})

orders["is_churned_90d"] = orders["days_since_last_order"] > 90
orders["is_active_30d"] = orders["days_since_last_order"] <= 30
orders[["customer_id", "days_since_last_order", "orders_last_90_days", "is_churned_90d", "is_active_30d"]]
```

## 5.2.6 Process diagrams reveal failure modes

### Why It Matters
A short data lineage sketch often uncovers duplicated records, missing time windows, or post-event information that would otherwise become leakage.


```python
import pandas as pd

events = pd.DataFrame({
    "customer_id": [1, 1, 2, 3],
    "event_value": [20, 30, 15, 50],
})
profiles = pd.DataFrame({
    "customer_id": [1, 1, 2, 3],
    "plan": ["basic", "plus", "plus", "enterprise"],
})

joined = events.merge(profiles, on="customer_id", how="left")
summary = {
    "raw_event_rows": len(events),
    "rows_after_join": len(joined),
    "duplicated_customer_ids_after_join": int(joined.duplicated(subset=["customer_id", "event_value"]).sum()),
}
summary
```
